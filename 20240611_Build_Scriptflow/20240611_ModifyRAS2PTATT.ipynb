```python
start = time.time()

# Configuration file path
configure_fpath = (
    r"D:\VENV_PYTHON\insar_scriptflow\B__post_insstackpsi_config.ini"
)

# ------------------------------------------------------------------------------
# Read and parse the configuration file
# ------------------------------------------------------------------------------
# Create a ConfigParser object to read the configuration file
config = configparser.ConfigParser()

# Read the content of the configuration file
config.read(configure_fpath)

# Print the path of the configuration file to confirm it's read correctly
print(f"Configuration file path: {configure_fpath}")

# ------------------------------------------------------------------------------
# Extract general parameters from the configuration file
# ------------------------------------------------------------------------------
# Main folder containing the INSSTACKPSI data
mainfolder = config["GENERAL"]["insstackpsi_mainfolder"]

# Path to the input M-file containing datetime information
input_mfile = config["GENERAL"]["input_mfile"]

# ------------------------------------------------------------------------------
# Extract RAS2PTATT specific parameters
# ------------------------------------------------------------------------------
# Column name for the x-coordinate in the displacement data
x_col = config["RAS2PTATT"]["x_col"]

# Column name for the y-coordinate in the displacement data
y_col = config["RAS2PTATT"]["y_col"]

# EPSG code for the coordinate reference system (CRS)
epsg = config["RAS2PTATT"]["coordinate_system"]

# ------------------------------------------------------------------------------
# Find the path to the raw cumulative displacement shapefile
# ------------------------------------------------------------------------------
# Use glob to find the shapefile matching the pattern "ras2pnt*RAW*.shp" in the main folder
raw_cdisp_fpath = glob(os.path.join(mainfolder, "ras2pnt*RAW*.shp"))[0]

# ------------------------------------------------------------------------------
# Extract datetime information from the input M-file
# ------------------------------------------------------------------------------
# Extract datetime strings from the input M-file using a helper function
mfile_datetimes = datetime_handle.extract_datetime_from_mfile(input_mfile)
mfile_datetimes = mfile_datetimes[1:]

# ------------------------------------------------------------------------------
# Convert the colname-duplicated shapefile to a normal DataFrame
# ------------------------------------------------------------------------------

modified_cumdisp = pd.read_pickle(
    r"E:\030_CHOUSHUI_2024\000_INSCALDEFO_2_INSSTACKPSI\PROCESS_003\ras2pnt_oChoushui_CUMDISP_LOS_mm.pkl",
    compression="zip",
)
# modified_cumdisp = dataframe_handle.modify_ras2ptatt_shp(
#     shp_fpath=raw_cdisp_fpath,
#     datetime_list=mfile_datetimes,
#     xcoord_colname=x_col,
#     ycoord_colname=y_col,
# )

modified_disp = dataframe_handle.cumdisp_to_disp_dataframe(
    cumdisp_dataframe=modified_cumdisp, datetime_list=mfile_datetimes
)

# ------------------------------------------------------------------------------
# Convert the DataFrame to a GeoDataFrame
# ------------------------------------------------------------------------------
# Convert the DataFrame to a GeoDataFrame using a helper function,
# specifying the x and y columns and the coordinate reference system
modified_cumdisp_geodf = convert_to_geodata(
    df=modified_cumdisp, xcoord_col=x_col, ycoord_col=y_col, crs_epsg=epsg
)

modified_disp_geodf = convert_to_geodata(
    df=modified_disp, xcoord_col=x_col, ycoord_col=y_col, crs_epsg=epsg
)



# Record the end time of the process
end = time.time()

# Calculate elapsed time in seconds, minutes, and hours
ellapse_time_seconds = round((end - start), 2)
ellapse_time_minutes = round((ellapse_time_seconds / 60), 2)
ellapse_time_hours = round((ellapse_time_seconds / 3600), 2)

# Log the total processing time
print("Processing time (seconds): " + str(ellapse_time_seconds))
print("Processing time (minutes): " + str(ellapse_time_minutes))
print("Processing time (hours): " + str(ellapse_time_hours))
```
