```python
import h5py

def read_hdf5_file(file_path):
    """
    Reads and displays information about an HDF5 file.
    
    Args:
        file_path (str): The path to the HDF5 file.
    """
    try:
        # Open the HDF5 file
        with h5py.File(file_path, 'r') as f:
            # Print the file name
            print(f"File: {file_path}")
            
            # Print the HDF5 file structure
            print("\nHDF5 File Structure:")
            def print_group(name, obj):
                if isinstance(obj, h5py.Dataset):
                    print(f"Dataset: {name}, shape: \[{', '.join(map(str, obj.shape))}\], type: {obj.dtype}")
                elif isinstance(obj, h5py.Group):
                    print(f"Group: {name}")
            f.visititems(print_group)
            
            # Print the attributes of the HDF5 file
            print("\nHDF5 File Attributes:")
            for attr_name, attr_value in f.attrs.items():
                print(f"{attr_name}: {attr_value}")
    except Exception as e:
        print(f"Error reading HDF5 file: {e}")

# Example usage
read_hdf5_file(fpath)
```
