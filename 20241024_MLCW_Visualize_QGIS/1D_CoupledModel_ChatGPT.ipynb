# Introduction to 1D Mechanical-Hydraulic Coupled Models

Understanding how fluids interact with solid materials is essential in many fields of engineering and science, such as civil engineering, geotechnics, and petroleum engineering. A **1D mechanical-hydraulic coupled model** is a mathematical representation that describes the interaction between mechanical behaviors (like stress and deformation in solids) and hydraulic behaviors (like fluid flow through a material) along a single spatial dimension.

This guide will help you grasp the fundamental concepts of this coupled model, explain why it is necessary, and compare it with traditional models that consider mechanical or hydraulic aspects separately.

---

## Basic Concepts

Before diving into the coupled model, let's review some foundational ideas in mechanics and hydraulics.

### Mechanics

**Mechanics** is the branch of physics that deals with the motion of objects and the forces that affect them. In the context of solid materials:

- **Stress**: The internal force per unit area within a material. Think of it as how much force is applied to a specific area.
- **Strain**: The deformation or displacement that occurs in a material due to applied stress.
- **Elasticity**: The ability of a material to return to its original shape after the stress is removed.

**Example**: When you compress a spring, it deforms (strain) due to the applied force (stress). Once you release it, it returns to its original shape because of its elasticity.

### Hydraulics

**Hydraulics** focuses on the behavior of fluids (liquids and gases) and how they flow through or around materials.

- **Fluid Pressure**: The force exerted by a fluid per unit area.
- **Permeability**: A measure of how easily a fluid can flow through a porous material.
- **Darcy's Law**: An equation that describes the flow of a fluid through a porous medium.

**Example**: Water seeping through soil is governed by hydraulic principles, where the soil's permeability affects how quickly the water moves.

---

## Limitations of Separate Mechanical and Hydraulic Models

### Mechanical-Only Models

Mechanical models that ignore fluid interactions can predict how a solid material deforms under load but fail to account for changes caused by fluid pressures within the material.

- **Limitation**: They cannot predict phenomena where fluid pressures significantly influence the material's mechanical behavior, such as soil consolidation due to water extraction.

### Hydraulic-Only Models

Hydraulic models that ignore the mechanical properties of the solid matrix can describe how fluids move through a material but do not consider how deformation of the material affects fluid flow.

- **Limitation**: They cannot accurately predict changes in flow paths due to deformation or failure of the solid structure, such as cracks forming in a dam.

## Why Do We Need a Coupled Model?

In many real-world situations, the mechanical and hydraulic behaviors are interdependent—they influence each other. Ignoring this coupling can lead to inaccurate predictions and potentially unsafe designs.

### Interdependence Between Mechanics and Hydraulics

   - Increased fluid pressure within a material can reduce the effective stress, potentially causing the material to deform or fail.
   - **Example**: Elevated pore water pressure in soil during heavy rainfall can lead to landslides.

2. **Mechanical Deformation Affects Fluid Flow**:

   - Deformation of the solid matrix can change the size and connectivity of pores, altering permeability and fluid flow paths.
   - **Example**: Compaction of soil due to loading reduces pore spaces, slowing down water drainage.

### Benefits of the Coupled Model

- **Accurate Predictions**: By considering both mechanical and hydraulic factors simultaneously, the coupled model provides more reliable predictions of material behavior.
- **Safety and Efficiency**: In engineering designs, accounting for coupled effects can prevent failures and optimize performance.
- **Understanding Complex Phenomena**: Some processes, like subsidence due to groundwater extraction, can only be explained using coupled models.

### Why Do We Need a Coupled Model?

Imagine you have a sponge filled with water. When you squeeze the sponge, water is pushed out, and the pressure of the water inside the sponge changes. Similarly, when you submerge the sponge in water, it absorbs the water, and its shape might change. This interplay between the sponge's structure and the movement of water within it is a basic example of **mechanical-hydraulic coupling**.

In many real-world scenarios, a similar interplay exists between the solid matrix (such as soil, rock, or biological tissue) and the fluid within it. Understanding this interaction is critical in fields like geomechanics, petroleum engineering, and environmental engineering. Here’s why a coupled model is necessary:

1. **Interdependent Behaviors**: 
   - **Mechanical changes** (like deformation or stress changes in a solid) can alter the **flow properties** of the fluid, such as permeability.
   - **Fluid flow** can cause changes in **mechanical properties**, like changing the pressure within a material, which can induce deformation.
   - For example, in groundwater flow through a soil layer, changes in pore water pressure can cause the soil to either compress or expand, affecting its permeability and structural integrity.

2. **Enhanced Predictive Power**: 
   - A purely **mechanical model** only considers how the solid structure behaves under stress but ignores how fluid flow might alter those stresses.
   - A purely **hydraulic model** focuses on how fluid moves through the material but doesn't account for how changes in the material's shape or stress might affect the flow.
   - The coupled model integrates both, leading to more accurate predictions of phenomena like **subsidence**, **landslides**, **wellbore stability**, or **compaction of reservoirs**.

### Basic Concepts: Mechanics, Hydraulics, and Their Coupling

#### 1. **Solid Mechanics**:
   - Focuses on how materials deform or stress when subjected to external forces.

     - **Stress ($\sigma$)**: Force per unit area within a material.

#### 2. **Hydraulic Flow**:
   - Describes how fluids move through a porous medium.
   - Governed by **Darcy's law** in porous media:
     $$
     q = -k \frac{dP}{dx}
     $$
     - $q$: Fluid flow rate.
     - $k$: Permeability of the medium.
     - $\frac{dP}{dx}$: Pressure gradient driving the flow.
   - Example: Determining how quickly water seeps through a layer of sand.

#### 3. **Coupled Mechanical-Hydraulic Model**:
   - Integrates the equations of solid mechanics and hydraulic flow.
   - Incorporates feedback loops between mechanical and hydraulic aspects:
     - **Mechanically induced fluid flow**: Changes in the solid structure (e.g., compression) can alter fluid flow by changing pore space.
     - **Fluid-induced mechanical changes**: Changes in fluid pressure can alter the stress distribution in the solid matrix, causing deformation.
   - Example: In a reservoir, as oil is extracted, the pressure decreases, causing the rock to compact and altering the permeability of the formation.

### How Does the 1D Coupled Model Work?

The 1D aspect of the model means we are only considering changes along one dimension (e.g., depth). This simplification makes the mathematics more manageable while still capturing the key interactions between fluid flow and solid deformation.

- **Mechanical Equation** (1D stress-strain relationship):
  $$
  \frac{d\sigma}{dz} + \rho g = 0
  $$
  - Where $\sigma$ is the stress, $z$ is the vertical depth, $\rho$ is the density, and $g$ is the gravitational acceleration.

- **Hydraulic Flow Equation** (1D Darcy’s law):
  $$
  q = -k \frac{dP}{dz}
  $$
  - Where $q$ is the flow rate and $P$ is the fluid pressure.

- **Coupling Equation**:
  - Changes in stress can alter permeability ($k$), and changes in fluid pressure ($P$) can affect the stress $\sigma$.

In practical applications, these equations are solved simultaneously to predict the changes in fluid pressure, stress, and deformation over time.

Certainly! Let's create an example of a 1D mechanical-hydraulic coupled model that simulates a simple soil consolidation problem due to fluid drainage over time. In this example, we will simulate how soil compresses under a constant load and how water drains out, reducing pore pressure and causing further compression. The example will be implemented using Python, with functions, detailed comments, and docstrings.

### Problem Statement

**Example**: A clay layer is subjected to a constant surface load, causing compression (mechanical behavior). Water within the pores drains over time (hydraulic behavior), reducing the pore pressure. The reduction in pore pressure changes the effective stress, which in turn affects the deformation of the clay layer. This process continues until equilibrium is reached.

We will simulate this using:
- 1D consolidation equations.
- A coupled relationship between pore pressure and effective stress.
- Time-dependent behavior of the system.

### Solution Approach

1. Define the **mechanical equation** using the effective stress principle.
2. Use **Darcy's law** to model fluid flow.
3. Solve the coupled equations using a time-stepping method (e.g., explicit Euler method).
4. Plot the results to visualize the change in pore pressure, effective stress, and settlement over time.

### Python Code

Let's implement the solution in Python, breaking down the logic into functions with detailed explanations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
THICKNESS = 10.0  # Thickness of the soil layer (meters)
TIME_STEPS = 1000  # Number of time steps
FINAL_TIME = 10000  # Total time for simulation (seconds)
DELTA_T = FINAL_TIME / TIME_STEPS  # Time step size (seconds)

# Mechanical properties
LOAD = 10000  # Applied surface load (Pa)
YOUNG_MODULUS = 1e7  # Young's modulus of the soil (Pa)
POISSON_RATIO = 0.3  # Poisson's ratio of the soil
COMPRESSIBILITY = 1 / YOUNG_MODULUS  # Compressibility of soil (1/Pa)

# Hydraulic properties
PERMEABILITY = 1e-8  # Permeability of the soil (m/s)
POROSITY = 0.3  # Porosity of the soil
WATER_VISCOSITY = 0.001  # Viscosity of water (Pa·s)
WATER_DENSITY = 1000  # Density of water (kg/m^3)

def compute_effective_stress(total_stress, pore_pressure):
    """
    Calculate the effective stress using the effective stress principle.
    
    Parameters:
        total_stress (float): Total applied stress (Pa).
        pore_pressure (float): Pore water pressure (Pa).
        
    Returns:
        float: Effective stress (Pa).
    """
    return total_stress - pore_pressure

def compute_pore_pressure_change(permeability, pore_pressure, compressibility, thickness, delta_t):
    """
    Calculate the change in pore pressure due to drainage (Darcy's law).
    
    Parameters:
        permeability (float): Permeability of the soil (m/s).
        pore_pressure (float): Current pore water pressure (Pa).
        compressibility (float): Compressibility of the soil (1/Pa).
        thickness (float): Thickness of the soil layer (m).
        delta_t (float): Time step (s).
        
    Returns:
        float: Change in pore pressure (Pa).
    """
    hydraulic_conductivity = permeability / (WATER_VISCOSITY * WATER_DENSITY)
    return -hydraulic_conductivity * pore_pressure * delta_t / (compressibility * thickness**2)

def simulate_consolidation(load, time_steps, delta_t, initial_pore_pressure):
    """
    Simulate the coupled mechanical-hydraulic consolidation process.
    
    Parameters:
        load (float): Applied surface load (Pa).
        time_steps (int): Number of time steps for the simulation.
        delta_t (float): Time step duration (s).
        initial_pore_pressure (float): Initial pore pressure (Pa).
        
    Returns:
        tuple: Arrays of time, pore pressure, effective stress, and settlement.
    """
    pore_pressure = initial_pore_pressure
    effective_stress = compute_effective_stress(load, pore_pressure)
    settlement = 0.0
    
    time_array = np.linspace(0, time_steps * delta_t, time_steps)
    pore_pressure_array = np.zeros(time_steps)
    effective_stress_array = np.zeros(time_steps)
    settlement_array = np.zeros(time_steps)
    
    for t in range(time_steps):
        # Store the current values
        pore_pressure_array[t] = pore_pressure
        effective_stress_array[t] = effective_stress
        settlement_array[t] = settlement
        
        # Compute change in pore pressure
        delta_pore_pressure = compute_pore_pressure_change(
            PERMEABILITY, pore_pressure, COMPRESSIBILITY, THICKNESS, delta_t
        )
        pore_pressure += delta_pore_pressure
        
        # Update effective stress based on new pore pressure
        effective_stress = compute_effective_stress(load, pore_pressure)
        
        # Compute settlement based on effective stress and soil compressibility
        settlement += effective_stress * COMPRESSIBILITY * delta_t
    
    return time_array, pore_pressure_array, effective_stress_array, settlement_array

# Initial conditions
initial_pore_pressure = 5000  # Initial pore pressure (Pa)

# Run the simulation
time_array, pore_pressure_array, effective_stress_array, settlement_array = simulate_consolidation(
    LOAD, TIME_STEPS, DELTA_T, initial_pore_pressure
)

# Plotting the results
plt.figure(figsize=(12, 8))

# Plot pore pressure over time
plt.subplot(3, 1, 1)
plt.plot(time_array, pore_pressure_array, label='Pore Pressure', color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Pore Pressure (Pa)')
plt.title('Pore Pressure over Time')
plt.grid(True)

# Plot effective stress over time
plt.subplot(3, 1, 2)
plt.plot(time_array, effective_stress_array, label='Effective Stress', color='green')
plt.xlabel('Time (s)')
plt.ylabel('Effective Stress (Pa)')
plt.title('Effective Stress over Time')
plt.grid(True)

# Plot settlement over time
plt.subplot(3, 1, 3)
plt.plot(time_array, settlement_array, label='Settlement', color='red')
plt.xlabel('Time (s)')
plt.ylabel('Settlement (m)')
plt.title('Settlement over Time')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Explanation of the Code

1. **Constants and Properties**: Define the soil layer properties, including thickness, mechanical properties like Young's modulus and compressibility, and hydraulic properties like permeability and porosity.

2. **Functions**:
   - `compute_effective_stress`: Calculates the effective stress as the difference between total stress and pore pressure.
   - `compute_pore_pressure_change`: Uses Darcy's law to calculate the change in pore pressure due to fluid flow.
   - `simulate_consolidation`: Runs the time-stepping simulation, updating pore pressure, effective stress, and settlement at each step.

3. **Simulation**:
   - Iteratively updates the pore pressure, effective stress, and settlement over each time step using an explicit time-stepping method.
   - Stores the results for plotting.

### Interpretation of the Results

- **$\frac{dx}{dt}$**: Rate of change at the current time step.

- Adjust the effective stress and settlement accordingly.

### 1. **Hydraulic Head and Water Flow**: How Water Moves

**Physical Meaning**:

In an aquifer, the water pressure is represented by something called the **hydraulic head** ($h$), which tells us the energy level of the groundwater.
- The change in hydraulic head over a distance causes water to flow, similar to how a steeper slope makes water flow faster.

**Mathematical Expression**:

**Physical Meaning**:
- Think about the weight of a book on a sponge filled with water. If the sponge is full, the water helps support the book’s weight. But if you squeeze out the water, the sponge has to bear more of the weight, so it compresses.

- **$\sigma'$**: Effective stress, the stress that the soil skeleton actually feels.

This equation tells us that the effective stress increases when the pore pressure decreases (as water is pumped out). It’s like saying, “As we squeeze out the water, the soil has to carry more of the weight on its own.”

**Physical Meaning**:

**Physical Meaning**:
- Settlement is the total sinking or downward movement of the ground surface. To calculate it, we need to add up all the tiny bits of compression (strains) across the entire thickness of the aquifer.

- **$S(t)$**: Settlement over time $t$.

```python
import numpy as np
import matplotlib.pyplot as plt

def initialize_simulation():
    """
    Initialize the simulation parameters and initial conditions.
    
    Returns:
        params (dict): Dictionary containing simulation parameters.
        h (numpy.ndarray): Initial hydraulic head array.
        sigma_prime (numpy.ndarray): Initial effective stress array.
        epsilon (numpy.ndarray): Initial cumulative strain array.
    """
    # Simulation parameters
    K = 1e-5          # Hydraulic conductivity (m/s)
    Ss = 1e-4         # Specific storage (1/m)
    E = 1e7           # Young's modulus (Pa)
    alpha = 0.8       # Biot’s coefficient
    H = 50.0          # Aquifer thickness (m)
    dt = 1000         # Time step size (s)
    total_time = 365 * 24 * 3600  # Total simulation time (s) - 1 year
    num_layers = 50   # Number of vertical discretization layers
    h0 = 20.0         # Initial hydraulic head (m)
    pumping_rate = -0.001  # Pumping rate at the top boundary (m/s)
    rho_w = 1000      # Density of water (kg/m³)
    g = 9.81          # Acceleration due to gravity (m/s²)
    sigma = rho_w * g * H  # Total stress (Pa)

    # Spatial discretization
    dz = H / num_layers               # Thickness of each layer (m)
    z = np.linspace(dz/2, H - dz/2, num_layers)  # Depth of each node (m)

    # Initialize arrays
    h = np.full(num_layers, h0)  # Hydraulic head array
    sigma_prime = np.full(num_layers, sigma - alpha * rho_w * g * h0)  # Effective stress array
    epsilon = np.zeros(num_layers)  # Cumulative strain array

    params = {
        'K': K,
        'Ss': Ss,
        'E': E,
        'alpha': alpha,
        'H': H,
        'dt': dt,
        'total_time': total_time,
        'num_layers': num_layers,
        'h0': h0,
        'pumping_rate': pumping_rate,
        'rho_w': rho_w,
        'g': g,
        'sigma': sigma,
        'dz': dz,
        'z': z,
        'time_steps': int(total_time / dt)
    }

    return params, h, sigma_prime, epsilon

def update_hydraulic_head_implicit(h, params):
    """
    Update the hydraulic head using the implicit finite difference method.
    
    Args:
        h (numpy.ndarray): Hydraulic head at current time step.
        params (dict): Dictionary containing simulation parameters.
    
    Returns:
        h_new (numpy.ndarray): Updated hydraulic head array.
    """
    K = params['K']
    Ss = params['Ss']
    dt = params['dt']
    dz = params['dz']
    num_layers = params['num_layers']
    pumping_rate = params['pumping_rate']
    
    # Coefficient for the finite difference
    r = K * dt / (Ss * dz **2)
    
    # Construct the tridiagonal matrix A
    main_diag = np.ones(num_layers) * (1 + 2 * r)
    lower_diag = np.ones(num_layers - 1) * (-r)
    upper_diag = np.ones(num_layers - 1) * (-r)
    
    # Adjust the main diagonal and right-hand side for boundary conditions
    b = h.copy()
    
    # Top boundary (i=0): specified flux (pumping rate)
    main_diag[0] = 1 + r
    upper_diag[0] = -r
    b[0] = h[0] + dt * pumping_rate / (Ss * dz)
    
    # Bottom boundary (i=N-1): no flow (impermeable)
    main_diag[-1] = 1 + r
    lower_diag[-1] = -r
    b[-1] = h[-1]
    
    # Assemble the tridiagonal matrix A
    A = np.diag(main_diag)
    A += np.diag(lower_diag, k=-1)
    A += np.diag(upper_diag, k=1)
    
    # Solve the linear system
    h_new = np.linalg.solve(A, b)
    
    return h_new

def update_effective_stress(h, params):
    """
    Update the effective stress based on the new hydraulic head.
    
    Args:
        h (numpy.ndarray): Hydraulic head array.
        params (dict): Dictionary containing simulation parameters.
    
    Returns:
        sigma_prime (numpy.ndarray): Updated effective stress array.
    """
    sigma = params['sigma']
    alpha = params['alpha']
    rho_w = params['rho_w']
    g = params['g']
    
    sigma_prime = sigma - alpha * rho_w * g * h
    
    return sigma_prime

def calculate_strain_and_settlement(delta_sigma_prime, epsilon, params):
    """
    Calculate incremental strain from change in effective stress and compute settlement.
    
    Args:
        delta_sigma_prime (numpy.ndarray): Change in effective stress array.
        epsilon (numpy.ndarray): Cumulative strain array.
        params (dict): Dictionary containing simulation parameters.
    
    Returns:
        epsilon (numpy.ndarray): Updated cumulative strain array.
        settlement (float): Total settlement at the ground surface.
    """
    E = params['E']
    dz = params['dz']
    
    # Calculate incremental strain
    delta_epsilon = delta_sigma_prime / E  # Assuming linear elastic behavior
    
    # Update cumulative strain
    epsilon += delta_epsilon
    
    # Compute settlement by integrating strain over depth
    settlement = np.sum(epsilon) * dz
    
    return epsilon, settlement

def run_simulation():
    """
    Run the time-stepping simulation to update and store results.
    
    Returns:
        time_array (numpy.ndarray): Array of time steps.
        h_history (list of numpy.ndarray): List of hydraulic head arrays over time.
        sigma_prime_history (list of numpy.ndarray): List of effective stress arrays over time.
        settlement_history (list of float): List of settlement values over time.
        params (dict): Dictionary containing simulation parameters.
    """
    # Initialize simulation
    params, h, sigma_prime, epsilon = initialize_simulation()
    num_steps = params['time_steps']
    dt = params['dt']
    
    # Arrays to store results
    time_array = np.arange(0, num_steps * dt, dt)
    h_history = []
    sigma_prime_history = []
    settlement_history = []
    
    # Initialize previous effective stress
    sigma_prime_prev = sigma_prime.copy()
    
    # Time-stepping loop
    for step in range(num_steps):
        # Update hydraulic head
        h = update_hydraulic_head_implicit(h, params)
        
        # Update effective stress
        sigma_prime = update_effective_stress(h, params)
        
        # Calculate change in effective stress
        delta_sigma_prime = sigma_prime - sigma_prime_prev
        
        # Calculate strain and settlement
        epsilon, settlement = calculate_strain_and_settlement(delta_sigma_prime, epsilon, params)
        
        # Update previous effective stress
        sigma_prime_prev = sigma_prime.copy()
        
        # Store results
        h_history.append(h.copy())
        sigma_prime_history.append(sigma_prime.copy())
        settlement_history.append(settlement)
        
        # Optional: Print progress every 100 steps
        if (step+1) % 100 == 0:
            print(f"Time step {step+1}/{num_steps} completed.")
    
    return time_array, h_history, sigma_prime_history, settlement_history, params

def plot_results(time_array, h_history, sigma_prime_history, settlement_history, params):
    """
    Plot the results of the simulation.
    
    Args:
        time_array (numpy.ndarray): Array of time steps.
        h_history (list of numpy.ndarray): List of hydraulic head arrays over time.
        sigma_prime_history (list of numpy.ndarray): List of effective stress arrays over time.
        settlement_history (list of float): List of settlement values over time.
        params (dict): Dictionary containing simulation parameters.
    """
    # Convert time to days for plotting
    time_days = time_array / (3600 * 24)
    
    # Plot hydraulic head at the top layer over time
    h_top = [h[0] for h in h_history]
    plt.figure(figsize=(10,6))
    plt.plot(time_days, h_top)
    plt.xlabel('Time (days)')
    plt.ylabel('Hydraulic Head at Top Layer (m)')
    plt.title('Hydraulic Head at Top Layer Over Time')
    plt.grid(True)
    plt.show()
    
    # Plot effective stress at the top layer over time
    sigma_prime_top = [sigma_prime[0] for sigma_prime in sigma_prime_history]
    plt.figure(figsize=(10,6))
    plt.plot(time_days, sigma_prime_top)
    plt.xlabel('Time (days)')
    plt.ylabel('Effective Stress at Top Layer (Pa)')
    plt.title('Effective Stress at Top Layer Over Time')
    plt.grid(True)
    plt.show()
    
    # Plot settlement over time
    plt.figure(figsize=(10,6))
    plt.plot(time_days, settlement_history)
    plt.xlabel('Time (days)')
    plt.ylabel('Settlement (m)')
    plt.title('Settlement Over Time')
    plt.grid(True)
    plt.show()

def main():
    """
    Main function to run the simulation and plot results.
    """
    # Run the simulation
    time_array, h_history, sigma_prime_history, settlement_history, params = run_simulation()
    
    # Plot the results
    plot_results(time_array, h_history, sigma_prime_history, settlement_history, params)

if __name__ == '__main__':
    main()
```

### 2. **Update Hydraulic Head Function**

**Function Name**: `update_hydraulic_head_implicit(h, params)`

**Purpose**: Updates the hydraulic head $ h $ over time using the implicit finite difference method for the 1D flow equation.

**Mathematical Notations and Concepts**:

- **Governing Equation**:

  The 1D transient groundwater flow equation, simplified by assuming constant total stress ($ \sigma $) and neglecting the coupling term, is:

  $$
  \frac{\partial}{\partial z} \left( K \frac{\partial h}{\partial z} \right) = S_s \frac{\partial h}{\partial t}
  $$

  Since $ K $ is constant, this simplifies to:

  $$
  K \frac{\partial^2 h}{\partial z^2} = S_s \frac{\partial h}{\partial t}
  $$

- **Finite Difference Approximation**:

  Discretize the spatial domain into $ N $ nodes with spacing $ \Delta z $. The second spatial derivative at node $ i $ is approximated by:

  $$
  \frac{\partial^2 h}{\partial z^2} \approx \frac{h_{i+1} - 2 h_i + h_{i-1}}{(\Delta z)^2}
  $$

  The time derivative is approximated by:

  $$
  \frac{\partial h}{\partial t} \approx \frac{h_i^{n+1} - h_i^n}{\Delta t}
  $$

  Substituting into the governing equation:

  $$
  K \frac{h_{i+1}^{n+1} - 2 h_i^{n+1} + h_{i-1}^{n+1}}{(\Delta z)^2} = S_s \frac{h_i^{n+1} - h_i^n}{\Delta t}
  $$

  Rearranged:

  $$
  -r h_{i-1}^{n+1} + (1 + 2r) h_i^{n+1} - r h_{i+1}^{n+1} = h_i^n
  $$

  where:

  $$
  r = \frac{K \Delta t}{S_s (\Delta z)^2}
  $$

- **Boundary Conditions**:

  - **Top Boundary (Node $ i = 0 $)**:

    The specified flux (pumping rate) at the top boundary is:

    $$
    -K \left. \frac{\partial h}{\partial z} \right|_{z=0} = q_{\text{pump}}
    $$

    Approximated using finite differences:

    $$
    -K \frac{h_1^{n+1} - h_0^{n+1}}{\Delta z} = q_{\text{pump}}
    $$

    Rearranged to include in the system:

    $$
    -r h_1^{n+1} + (1 + r) h_0^{n+1} = h_0^n + \frac{q_{\text{pump}} \Delta t}{S_s \Delta z}
    $$

  - **Bottom Boundary (Node $ i = N - 1 $)**:

    No-flow boundary condition:

    $$
    \left. \frac{\partial h}{\partial z} \right|_{z=H} = 0 \implies h_{N}^{n+1} = h_{N-1}^{n+1}
    $$

    In finite difference form:

    $$
    -r h_{N-2}^{n+1} + (1 + r) h_{N-1}^{n+1} = h_{N-1}^n
    $$

- **System of Equations**:

  The discretized equations form a tridiagonal system that can be represented as:

  $$
  \mathbf{A} \mathbf{h}^{n+1} = \mathbf{b}
  $$

  where:

  - $ \mathbf{A} $ is a tridiagonal matrix incorporating coefficients from the discretized equations.
  - $ \mathbf{h}^{n+1} $ is the vector of hydraulic heads at the new time step.
  - $ \mathbf{b} $ is the right-hand side vector containing known values from the previous time step and boundary conditions.

- **Solution**:

  The system is solved for $ \mathbf{h}^{n+1} $ using linear algebra methods (e.g., Thomas algorithm for tridiagonal systems).
