```python
dir_1 = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2023_葉 大綱\2023_NEW_20240515\20240527_PROCESSED_DATA\_2022\1__Datetime_Modified"
dir_2 = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2023_葉 大綱\2023_NEW_20240515\20240527_PROCESSED_DATA\_2023\1__Datetime_Modified"
```

```python
files_dir_1 = os.listdir(dir_1)
files_dir_2 = os.listdir(dir_2)

print(len(files_dir_1), len(files_dir_2))
similar_files = sorted(list(set(files_dir_1).intersection(set(files_dir_2))))
single_files = [file for file in files_dir_1 if file not in similar_files]
single_files
```

```python
savefolder = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2023_葉 大綱\2023_NEW_20240515\20240527_PROCESSED_DATA\_MERGED\1__Datetime_Modified"
savename = "GPS_2022_2023_merged_mod.xlsx"
savepath = os.path.join(savefolder, savename)
savepath
```

```python
for select_file in tqdm(single_files):
    station_name = select_file.split("_")[0]
    available_dirs = [dir_1, dir_2]

    _temp = pd.DataFrame(data=None, index=None)

    for dirname in available_dirs:
        fpath = os.path.join(dirname, select_file)
        if os.path.isfile(fpath):
            df = pd.read_excel(fpath)
            df.columns = [col.replace("(m)", "(cm)") for col in df.columns]
            _temp = pd.concat([_temp, df], ignore_index=True)
        else:
            print(f"{fpath} not exist")
            pass

    data_io.save_df_to_excel(df_to_save=_temp, filepath=savepath, sheet_name=station_name, verbose=False)
```
