```python
def change_first_colname(input_df, by_value="temp"):
    return input_df.rename({input_df.columns[0]:by_value}, axis=1)
```

```python
fpath = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2023_葉 大綱\2023_NEW_20240515\2023_GNSS.xlsx"
processing_year = int(os.path.basename(fpath).split("_")[0])
available_sheetnames = data_io.get_sheetnames(fpath)
print("Year to process:", processing_year)
```

```python
target_savefolder = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2023_葉 大綱\2023_NEW_20240515\20240527_PROCESSED_DATA\_{}\Datetime_Modified".format(processing_year)
```

```python
for select_sheetname in available_sheetnames:
    try:
        df = pd.read_excel(fpath, sheet_name=select_sheetname)
        
        df = change_first_colname(df, by_value="Datetime")
        
        df["Datetime"] = pd.date_range(start=datetime(processing_year, 1, 1), end=datetime(processing_year, 12, 31))
        df.set_index("Datetime", inplace=True)
        savename = f"{select_sheetname}_mod.xlsx"
        savepath = os.path.join(target_savefolder, savename)
        df.to_excel(savepath)
    except Except as e:
        print(f"{e} at {select_sheetname}")
        pass
```
