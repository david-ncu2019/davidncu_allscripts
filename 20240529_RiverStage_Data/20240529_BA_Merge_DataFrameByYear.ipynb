```python
def file_by_station(folder2search, station_name):
    temp = []
    for root, dirs, files in os.walk(folder2search):
        for filename in fnmatch.filter(files, f"{station_name}*"):
            temp.append(os.path.join(root, filename))
    return temp
```

```python
mainfolder = r"D:\002_DATA\__RIVER_STAGE\PROCESSED_DATA_byYEAR"
savefolder = r"D:\002_DATA\__RIVER_STAGE\MERGED_DATA"
```

```python
info_df = pd.read_excel("Output_File_Info_Summary.xlsx")
info_df.head(5), info_df.shape
```

```python
station_name_unique = info_df["STATION"].unique()
json_fpath = r"D:\002_DATA\__RIVER_STAGE\PROCESSED_DATA_byYEAR\fpath_byStationName.json"
fpath_byStationName_dict = data_io.read_json_to_dict(json_fpath)

# select_station = station_name_unique[0]
for select_station in tqdm(station_name_unique[:], desc="Stations", position=0, ncols=500):

    fpath_list_byStationName = fpath_byStationName_dict[select_station]
    
    region, basin, tributary = fpath_list_byStationName[0].split("\\")[4:7]
    
    merged_df = pd.concat([pd.read_excel(fpath) for fpath in fpath_list_byStationName], ignore_index=True)
    merged_df = merged_df.sort_values(by="time")
    
    output_savefolder = os.path.join(savefolder, region, basin, tributary)
    if not os.path.exists(output_savefolder):
        os.makedirs(output_savefolder)
    
    merged_df.to_excel(os.path.join(output_savefolder, f"{select_station}.xlsx"), index=False)
```
