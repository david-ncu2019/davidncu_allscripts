```python
from appgeopy import *
from my_packages import *
from tqdm.notebook import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
mainfolder = r"D:\002_DATA\__RIVER_STAGE\MERGED_DATA"
plt.rcParams["font.sans-serif"] = ["SimHei"]

all_files = glob(os.path.join(mainfolder, "*", "*", "*", "*.xlsx"))

select_fpath = all_files[0]

for select_fpath in tqdm(all_files, position=0, ncols=500):

    # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # 
    
    region, basin, tributary, station = select_fpath.split("\\")[-4:]
    
    df = pd.read_excel(select_fpath, parse_dates=[0], index_col=[0])
    
    # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
    fig_title = "Basin: {} - Tributary: {} - Station: {}".format(
        basin, tributary, station.split(".")[0]
    )
    
    output_fig, ax = visualize.plot_single_graph(
        df,
        title=fig_title,
        xlabel="",
        ylabel="Daily Mean Level (m)",
        line_label="Daily Mean Level (m)",
        figsize=(11.7, 8.27 * 2 / 3),
        linestyle="--",
        marker="o",
        color="blue",
        alpha=0.3,
    )
    
    years = mdates.YearLocator(
        base=1,
    )
    months = mdates.MonthLocator(bymonth=[3, 5, 7, 9, 11, 1])
    monthyearFmt = mdates.DateFormatter("%Y/%m")
    
    ax.xaxis.set_major_locator(years)
    ax.xaxis.set_minor_locator(months)
    
    ax.grid(which="major", color="grey", axis="x")
    ax.grid(which="minor", color="lightgrey", axis="x")
    ax.set_axisbelow(True)
    
    start_time = datetime(2017, 12, 1) - timedelta(days=15)
    end_time = datetime(2022, 12, 31) + timedelta(days=15)
    ax.set_xlim(start_time, end_time)
    
    fig_savename = "{}_{}_{}.png".format(basin, tributary, station.split(".")[0])
    
    fig_savefolder = os.path.join(r"D:\002_DATA\__RIVER_STAGE\DATA_FIGURES", region.upper())
    if not os.path.exists(fig_savefolder):
        os.makedirs(fig_savefolder)
    
    visualize.save_figure(
        output_fig, savepath=os.path.join(fig_savefolder, fig_savename), dpi=150
    )

    plt.close()
```
