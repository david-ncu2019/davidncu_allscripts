```python
string_decoder = lambda x: x.decode("utf-8")
```

```python
start_time = datetime(year=2000, month=1, day=1)
end_time = datetime(year=2025, month=1, day=1)

# Create a date range with monthly frequency
date_index = pd.date_range(start=start_time, end=end_time, freq="MS")

output_df = pd.DataFrame(data=None, index=date_index)

# select_station = available_stations[0]
for select_station in tqdm(available_stations):
    monthly_datetime = pd.to_datetime(
        [string_decoder(ele) for ele in mlcw_measure[select_station]["monthly_date"]]
    )
    monthly_compact_byLayer = mlcw_measure[select_station]["monthly_values"]["compactbylayer"]
    temp = pd.Series(monthly_compact_byLayer[0], index=monthly_datetime)
    output_df[select_station] = output_df.index.map(temp)

output_df = output_df.T
output_df["geometry"] = output_df.index.map(mlcw_gdf["geometry"])
output_gdf = gpd.GeoDataFrame(output_df, geometry=output_df["geometry"])
# output_gdf = output_gdf.dropna(axis=1, how="all")

new_cols = [
    ele.strftime("%Y%m%d") if not isinstance(ele, str) else ele for ele in output_gdf.columns
]
output_gdf.columns = new_cols
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# Load and prepare data
cache_df = pd.read_csv("time_and_station_number.csv")
cache_df['time'] = pd.to_datetime(cache_df['time'], errors='coerce')
cache_df = cache_df.dropna(subset=['time'])
cache_df = cache_df.sort_values("time")

# Calculate appropriate bar width based on data frequency
time_diff = cache_df['time'].diff().median()
width_days = max(2, time_diff.days)  # Minimum 2-day width for visibility
# width_days = 15
width = width_days  # Width in days

# Create visualization
fig, ax = plt.subplots(figsize=(12, 6))

# Generate bar chart with appropriate width
ax.bar(cache_df['time'], cache_df['num_stations'], 
       width=width, color='darkgreen', alpha=0.5)

# Configure time axis
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_major_locator(mdates.YearLocator(2))  # Every 2 years
ax.xaxis.set_minor_locator(mdates.YearLocator())  # Minor ticks every year

# Force x-axis limits to match data range with padding
date_range = cache_df['time'].max() - cache_df['time'].min()
padding = date_range * 0.02  # 2% padding on each side
ax.set_xlim(cache_df['time'].min() - padding, 
            cache_df['time'].max() + padding)

# Styling
plt.xticks(rotation=45, ha='right')
ax.set_ylabel('Number of Stations', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# # Print sample data points for verification
# print(f"First 5 data points:\n{cache_df.head()}")

# # Add data presence indicator
# if len(cache_df) > 0:
#     ax.text(0.5, 0.95, f"Total data points: {len(cache_df)}", 
#             transform=ax.transAxes, ha='center')

fig.tight_layout()
visualize.save_figure(fig=fig, savepath="available_data_byMonths.png")
plt.show()
```
