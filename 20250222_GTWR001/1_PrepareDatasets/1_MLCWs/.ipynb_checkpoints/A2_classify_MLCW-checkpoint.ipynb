```python
key2search = "LandSubsidenceMonitoringWellIdentifier"
file_folder = r"E:\SUBSIDENCE_PROJECT_DATA\2024_MLCW_ANALYSIS\megnaticRingsLayer"

# select_station = available_stations[0]
for select_station in tqdm(available_stations):

    station_code = mlcw_metadata[select_station][key2search]
    filename2search = station_code.split("_")[-1] + "*.xlsx"
    
    found_files = glob(os.path.join(file_folder, filename2search))
    
    if found_files:
        file2open = found_files[0]
    else:
        print(f"{select_station} {station_code},  File not found")
        # break
    
    df = pd.read_excel(file2open)
    colnames = df.columns.tolist()
    
    classify_dict = {"classify": {"depth": df["depth(m)"].tolist(), "layer": df["Layer"].tolist()}}
    mlcw_data[select_station].update(classify_dict)
```
