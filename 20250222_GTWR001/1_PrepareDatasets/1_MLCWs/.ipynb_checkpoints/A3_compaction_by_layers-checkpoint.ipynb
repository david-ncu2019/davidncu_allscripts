```python
h5_fpath = "20250313_MLCW_CRFP_v10.h5"
# initiate MLCW class object
mlcw_obj = MLCW(h5_fpath=h5_fpath)
mlcw_data, mlcw_metadata = mlcw_obj.get_data()
available_stations = mlcw_obj.list_stations()
available_stations[:5]
```

```python
mlcw_obj.list_structure(mlcw_data)
```

```python
# select_station = available_stations[0]
# print(f"Station: {select_station}")

for select_station in tqdm(available_stations):

    # extract compaction ring by ring from the corresponding station
    mlcw_ringbyring = mlcw_obj.build_dataframe(station=select_station, value_type="ringbyring")

    # extract measurement data from station
    data_byStation = mlcw_data[select_station]

    classify_info = pd.DataFrame(data_byStation["classify"])
    # Decode byte strings in 'layer' column
    classify_info["layer"] = classify_info["layer"].apply(lambda x: x.decode("utf-8") if isinstance(x, bytes) else x)

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    sum_comp_allLayer = pd.DataFrame(data=None)

    # - - - - loop - - - -

    # layer2process = sorted(set([ele for ele in classify_info["layer"].tolist() if len(ele) <= 2]))

    layer_order = 1
    # for layer_order in ["1", "2", "3"]:
    for filter_cond in [["F1", "T1"], ["F2", "T2"], ["F3", "T3"], ["F3隞乩?"]]:

        # filter_cond = [ele for ele in layer2process if layer_order in ele]

        # extract info by select layer
        classify_info_byLayer = classify_info[classify_info["layer"].isin(filter_cond)]
        if len(classify_info_byLayer) > 0:

            start_depth = classify_info_byLayer.iloc[0, 0]
            end_depth = classify_info_byLayer.iloc[-1, 0]

            mlcw_byLayer = mlcw_ringbyring.query("index>=@start_depth & index<=@end_depth")

            sum_comp_byLayer = mlcw_byLayer.sum(axis=0)

            sum_comp_byLayer.name = "L" + str(layer_order)

            sum_comp_allLayer = pd.concat([sum_comp_allLayer, sum_comp_byLayer], axis=1)

        layer_order += 1

    data_io.save_df_to_excel(
        df_to_save=sum_comp_allLayer,
        filepath="compact_by_layers.xlsx",
        sheet_name=select_station,
        verbose=False,
        index=True,
    )
    mlcw_data[select_station]["values"]["compactbylayer"] = sum_comp_allLayer.T.to_numpy()

# --------- plotting to compare -------------------------------------
# fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(11.7, 8.3*2/3))
# visualize.base_plot(
#     ax=ax,
#     data=sum_comp_allLayer.sum(axis=1),
#     label="Three Layers Cumulative Compaction",
#     title=select_station,
#     marker="o",
#     linestyle="--",
#     color="gray",
#     alpha=1
# )
# visualize.base_plot(
#     ax=ax,
#     data=mlcw_ringbyring.sum(axis=0),
#     label="All Ring Cumulative Compaction",
#     title=select_station,
#     marker="",
#     linestyle="--",
#     color="blue",
#     lw=3,
#     alpha=1
# )
# visualize.configure_axis(ax=ax, tick_direction="out", hide_spines=["top", "right"])
# visualize.configure_legend(ax=ax, fontsize_base=14)
# visualize.configure_datetime_ticks(ax=ax, grid=True)
# # visualize.save_figure(fig=fig, savepath=f"figs_ringbyring\\{select_station}_compare.png")
# plt.close()
```
