Assign the backtransformed CUMDISP values at grid points to the GTWR regression points, so we can connect the model coefficients and the predictor values --> prediction values

```python
backtransform_disp_df = pd.read_pickle(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\5_PostKriging\PostKriging_Backtransform_Data\grid_points\BACKTRANSFORMED_Monthly_DISPLACEMENT_dU_CRFP_Full.xz"
)
```

```python
# convert InSAR backtransformed disp to cumdisp
disp_info_df = backtransform_disp_df.iloc[:, :3]
disp_measure = backtransform_disp_df.iloc[:, 3:]

backtransform_cumdisp_df = pd.concat([disp_info_df, disp_measure.cumsum(axis=1)], axis=1)

backtransform_cumdisp_df.head(5)
```

**2025/4/23**

I successfully extract the GTWR coeffients into all grid points, using chunking technique.

Previously, I cannot perform it because at each time period, I include too many points.

I found that with 5000 points for each time period, the algorithm in R program work well.

Increase to 10_000 points, it will take forever to finish the progress.

```python
# get the current working directory
cwd = os.getcwd()

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
# only get the measurement values
# of the cumulative displacement table
backtransform_cumdisp_measurement = backtransform_cumdisp_df.iloc[:, 2:]

# extract the measurement columns for later use
cumdisp_columns = [col for col in backtransform_cumdisp_measurement.columns if col.startswith("N")]

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 

# get all GTWR coeffient files
gtwr_coeff_files = glob(os.path.join(cwd, "*", "*", "gtwr*bisquare*Layer*.xz"))

# select one file to work with
# select_gtwr_coeff = gtwr_coeff_files[0]
for select_gtwr_coeff in tqdm(gtwr_coeff_files, desc="GTWR by Layer"):
    
    file_basename = os.path.basename(select_gtwr_coeff).split(".")[0]
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
    # create an empty table to store results after adding information
    gtwr_coeff_full = pd.DataFrame(data=None)
    
    # read file
    gtwr_coeff_df = pd.read_pickle(select_gtwr_coeff)
    
    available_times = [f"t{idx}" for idx in range(len(cumdisp_columns))]
    
    idx = 0  # the number of indexes = number of columns
    for idx in trange(len(available_times), desc="Time Period", position=1, leave=False):
        
        # extract the coeffients corresponding to the current time period
        gtwr_coeff_byTimePeriod = gtwr_coeff_df.query("time_period==@idx")
        
        # rename the column of coeffients
        gtwr_coeff_byTimePeriod = gtwr_coeff_byTimePeriod.rename({"CUMDISP": "CUMDISP_coeff"}, axis=1)
        
        # set PointKey as index
        # to assign the cumulative displacements corresponding to points
        gtwr_coeff_byTimePeriod = gtwr_coeff_byTimePeriod.set_index("PointKey")
        # select the corresponding measurement column
        cumdisp_measurement_byTime = backtransform_cumdisp_measurement.iloc[:, idx]
        
        gtwr_coeff_byTimePeriod["CUMDISP_measure"] = gtwr_coeff_byTimePeriod.index.map(cumdisp_measurement_byTime)
        
        # calculate the predicted values based on the coeffients
        gtwr_coeff_byTimePeriod["MLCW_pred"] = (
            gtwr_coeff_byTimePeriod["CUMDISP_measure"] * gtwr_coeff_byTimePeriod["CUMDISP_coeff"] + gtwr_coeff_byTimePeriod["Intercept"]
        )
        
        gtwr_coeff_full = pd.concat([gtwr_coeff_full, gtwr_coeff_byTimePeriod])
    
    scaler = StandardScaler()
    gtwr_coeff_full["Intercept_scaled"] = scaler.fit_transform(
        gtwr_coeff_full[["Intercept"]]
    )
    gtwr_coeff_full["CUMDISP_coeff_scaled"] = scaler.fit_transform(
        gtwr_coeff_full[["CUMDISP_coeff"]]
    )
    
    gtwr_coeff_full = gtwr_coeff_full.loc[:, sorted(gtwr_coeff_full.columns)]
    gtwr_coeff_full.to_pickle(f"{file_basename.upper()}_Full.xz")
```
