```python
backtransform_disp_df = pd.read_pickle(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\5_PostKriging\PostKriging_Backtransform_Data\grid_points\BACKTRANSFORMED_Monthly_DISPLACEMENT_dU_CRFP_Full.xz"
)

# convert InSAR backtransformed disp to cumdisp
disp_info_df = backtransform_disp_df.iloc[:, :3]
disp_measure = backtransform_disp_df.iloc[:, 3:]

backtransform_cumdisp_df = pd.concat([disp_info_df, disp_measure.cumsum(axis=1)], axis=1)

measure_cols = [col for col in backtransform_cumdisp_df.columns if col.startswith("N")]

show(backtransform_cumdisp_df)
```

```python
regpoints_files = glob(os.path.join("regpoints_filefld", "*.feather"))

# select_file = regpoints_files[0]
for select_file in tqdm(regpoints_files):

    basename = os.path.basename(select_file)
    savepath = os.path.join("regpoints_filefld_addCUMDISP", basename)

    df = pd.read_feather(select_file)
    df = df.set_index("PointKey")

    monthly_intervals = df["monthly"].unique()
    
    updated_df = pd.DataFrame(data=None, index=None)
    
    # idx = 0
    for idx in trange(len(monthly_intervals), position=1, leave=False):
        current_month_interval = monthly_intervals[idx]
        df_byTimePeriod = df.query("monthly==@current_month_interval")
        df_byTimePeriod["CUMDISP"] = df_byTimePeriod.index.map(backtransform_cumdisp_df.loc[:, measure_cols[idx]])
    
        updated_df = pd.concat([updated_df, df_byTimePeriod])
    
    updated_df.reset_index(drop=False)
    updated_df.to_feather(savepath)
    
```
