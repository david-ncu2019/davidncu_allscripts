```python
gtwr_flds = glob("gtwr_Layer*")
gtwr_flds
```

```python
# select_fld = gtwr_flds[0]
for select_fld in tqdm(gtwr_flds):

    gtwr_coeff_output_df = pd.DataFrame(data=None, index=None)
    
    chunk_flds = glob(os.path.join(select_fld, "grid_pnt_chunk*"))
    for select_chunk_fld in tqdm(chunk_flds, leave=True):
        # select_chunk_fld = chunk_flds[0]
        csv_files = glob(os.path.join(select_chunk_fld, "*csv"))
        df_byChunk = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)
        pointkey = [
            f"X{int(x_twd97*1000)}Y{int(y_twd97*1000)}"
            for x_twd97, y_twd97 in zip(df_byChunk["X_TWD97"], df_byChunk["Y_TWD97"])
        ]
        df_byChunk.insert(loc=0, column="PointKey", value=pointkey)
        gtwr_coeff_output_df = pd.concat([gtwr_coeff_output_df, df_byChunk], ignore_index=True) 
    
    gtwr_coeff_output_df = gtwr_coeff_output_df.sort_values(by=["time_period", "PointKey"]).reset_index(drop=True)
    gtwr_coeff_output_df.to_pickle(f"{select_fld}\\{select_fld}_coeffs_regpoints.xz")
```
