```python
from mgtwr.sel import SearchMGTWRParameter
from mgtwr.model import MGTWR
import pandas as pd
```

```python
data = pd.read_csv("example.csv")
coords = data[['longitude', 'latitude']]
t = data[['t']]
X = data[['x1', 'x2']]
y = data[['y']]
```

```python
coords.shape
```

`bws` tuple description:
1. **`numpy.ndarray`, length 3**: Optimized spatial bandwidths (`opt_bws`).
2. **`numpy.ndarray`, length 3**: Optimized temporal bandwidths (`opt_tau`).
3. **`numpy.ndarray`, length 8**: History of spatial bandwidths over iterations (`np.array(BWs)`).
4. **`numpy.ndarray`, length 8**: History of temporal bandwidths over iterations (`np.array(Taus)`).
5. **`numpy.ndarray`, length 8**: Convergence scores (`np.array(scores)`).
6. **`numpy.ndarray`, length 1728**: Final regression coefficients for each observation and covariate (`Betas`).
7. **`numpy.ndarray`, length 1728**: Final residuals (`err`).
8. **`numpy.float64`**: Initial GTWR spatial bandwidth (`bw_gtwr`).
9. **`numpy.float64`**: Initial GTWR temporal bandwidth (`tau_gtwr`).

```python
sel_multi = SearchMGTWRParameter(coords, t, X, y, kernel='gaussian', fixed=True)
bws = sel_multi.search(multi_bw_min=[0.1], verbose=True, tol_multi=1.0e-4, time_cost=True)
```
