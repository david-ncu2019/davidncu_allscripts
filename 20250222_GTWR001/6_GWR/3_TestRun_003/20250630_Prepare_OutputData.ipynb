```python
top_folder = "GWR_Prediction_Output/"
output_folder_byLayer = glob(os.path.join(top_folder, "Layer*"))
output_folder_byLayer
```

```python
# # select_folder = output_folder_byLayer[0]
for select_folder in tqdm(output_folder_byLayer):
    layer_name = os.path.basename(select_folder)
    output_df = pd.DataFrame(data=None, index=None)
    
    for i in trange(1, 67):
    # i = 1
        search_key= f"*_time{i}_*"
        found_file = glob(os.path.join(select_folder, search_key))[0]
        df = pd.read_csv(found_file)
        df = df.drop(labels=["X_TWD97.1", "Y_TWD97.1"], axis=1)
        df["prediction_std"] = np.sqrt(df["prediction_var"])
        df["monthly"] = i
        pointkey_arr = [f"X{int(x*1000)}Y{int(y*1000)}" for x,y in zip(df["X_TWD97"], df["Y_TWD97"])]
        df.insert(loc=0, column="PointKey", value=pointkey_arr)
        output_df = pd.concat([output_df, df], ignore_index=True)
    
    output_df.to_feather(os.path.join(top_folder, f"gwr_prediction_{layer_name}.feather"))
```
