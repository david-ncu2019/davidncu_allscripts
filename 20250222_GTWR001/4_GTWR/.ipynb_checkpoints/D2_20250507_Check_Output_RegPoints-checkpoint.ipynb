```python
mlcw_gdf = gpd.read_file(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\points_fld\mlcw_twd97.shp"
)
mlcw_gdf.head(5)
```

```python
csv_files = glob("1_TestRun_101/gtwr*.csv")
csv_files[:]
```

```python
# select_csv = csv_files[1]
for select_csv in tqdm(csv_files):

    # select_csv = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\4_GTWR\1_TestRun_101\gtwr_coefficients_exponential_Layer_1~CUMDISP_lambda0d05.csv"
    
    basename = os.path.basename(select_csv)
    
    layer_from_basename = "_".join(basename.split("_")[3:5])
    model_name = basename.split("_")[2]
    lambda_value = basename.split(".")[0].split("_")[-1]
    
    df = pd.read_csv(select_csv)
    
    df["PointKey"] = [f"X{int(x*1000)}Y{int(y*1000)}" for x, y in zip(df["X_TWD97"], df["Y_TWD97"])]
    unique_pointkeys = df["PointKey"].unique()
    
    # select_pointkey = unique_pointkeys[0]
    for select_pointkey in tqdm(unique_pointkeys, position=1, leave=False):
    
        mlcw_station = mlcw_gdf.query("PointKey==@select_pointkey").STATION.values[0]
    
        df_byPointKey = df.query("PointKey==@select_pointkey")
    
        time_arr = df_byPointKey["time_stamp"]
        original_vals = df_byPointKey["y"]
        predicted_vals = df_byPointKey["yhat"]
        residuals = df_byPointKey["residual"]
        slope_vals = df_byPointKey["CUMDISP"]
        intercept_vals = df_byPointKey["Intercept"]
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(8.3, 11.7), sharex=True)
        axes = axes.flatten()
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        ax1 = axes[0]
        ax1.plot(
            time_arr, original_vals, marker="o", linestyle="--", color="black", lw=1.5, label="Original"
        )
        ax1.plot(
            time_arr,
            predicted_vals,
            marker="o",
            linestyle="--",
            color="orangered",
            lw=1.5,
            label="Predicted",
        )
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        ax2 = axes[1]
        ax2.plot(
            time_arr,
            residuals,
            marker="o",
            linestyle="--",
            color="darkgreen",
            lw=1.5,
            label="Residuals",
        )
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        ax3 = axes[2]
        ax3.plot(time_arr, slope_vals, marker="o", linestyle="--", color="blue", lw=1.5, label="Slopes")
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        ax4 = axes[3]
        ax4.plot(
            time_arr,
            intercept_vals,
            marker="o",
            linestyle="--",
            color="blueviolet",
            lw=1.5,
            label="Intercepts",
        )
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        fig.suptitle(f"{mlcw_station}    {layer_from_basename}", fontsize=16, fontweight="bold")
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        for ax in axes:
            visualize.configure_axis(ax=ax, hide_spines=["right", "top"], tick_direction="out")
            visualize.configure_legend(ax=ax, loc="best", fontsize_base=14)
            visualize.configure_ticks(ax=ax, x_major_interval=5)
            ax.grid()
        # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
        fig.tight_layout()
    
        fig_savefolder = os.path.join(
            "out_figs/",
            os.path.basename(os.path.dirname(select_csv)),
            f"{model_name}_{layer_from_basename}_{lambda_value}",
        )
    
        if not os.path.exists(fig_savefolder):
            os.makedirs(fig_savefolder)
    
        visualize.save_figure(fig, savepath=os.path.join(fig_savefolder, f"{mlcw_station}.png"))
        plt.close()
```
