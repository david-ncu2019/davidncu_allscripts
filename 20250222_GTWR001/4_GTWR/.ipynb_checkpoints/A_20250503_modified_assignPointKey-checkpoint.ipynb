```python
coeffs_files = glob(r"5_TestRun_105/*/bisquare*coeffs*.feather")
coeffs_files
```

```python
# select_file = coeffs_files[0]

for select_file in tqdm(coeffs_files):

    dirname = os.path.dirname(select_file)
    basename = os.path.basename(select_file)
    
    gtwr_coeff_output_df = pd.DataFrame(data=None, index=None)
    
    feather_df = pd.read_feather(select_file)
    
    pointkey = [
        f"X{int(x_twd97*1000)}Y{int(y_twd97*1000)}"
        for x_twd97, y_twd97 in zip(feather_df["X_TWD97"], feather_df["Y_TWD97"])
    ]
    
    feather_df.insert(loc=0, column="PointKey", value=pointkey)
    
    feather_df = feather_df.sort_values(by=["time_period", "PointKey"]).reset_index(drop=True)
    
    feather_df.to_pickle(os.path.join(dirname, basename.replace(".feather", ".xz")))
```
