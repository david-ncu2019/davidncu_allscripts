This code snippet helps me to search for the closest GPS station to the MLCW station

This code is used to search for the MLCW and GPS contains the longest time series

```python
mlcw_gps_distance_table = pd.read_excel("MLCW_and_Nearest_GPS.xlsx")
mlcw_gps_distance_table.head(5)
```

```python
gps_folder = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\4_GTWR\PCA_Imputed"
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
mlcw_fpath = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\1_PrepareDatasets\1_MLCWs\20250415_MLCW_CRFP_monthly_v2.h5"
mlcw_obj = MLCW(mlcw_fpath)
mlcw_measures, mlcw_metadata = mlcw_obj.get_data()
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
available_stations = mlcw_gps_distance_table["MLCW_Station"].tolist()
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

cache = {
    "MLCW_Station": [],
    "GPS_Station": [],
    "MLCW_Start": [],
    "GPS_Start": [],
    "MLCW_End": [],
    "GPS_End": [],
    "Valid_Start": [],
    "Valid_End": [],
}

# select_MLCW_station = available_stations[0]
for select_MLCW_station in available_stations:
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    # define a function for string decoder
    string_decoder_func = lambda arr: [x.decode("utf-8") for x in arr]

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    measures_byStation = mlcw_measures[select_MLCW_station]
    monthly_date_arr = pd.to_datetime(string_decoder_func(measures_byStation["monthly_date"]))
    monthly_values_arr = measures_byStation["monthly_values"]["compactbylayer_PCA"]
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    cdisp_mlcw_df = pd.DataFrame(data={"time": monthly_date_arr})

    n_layers = monthly_values_arr.shape[0]

    for i in range(n_layers):
        cdisp_mlcw_df[f"Layer_{i+1}"] = monthly_values_arr[i]

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    cdisp_mlcw_df = cdisp_mlcw_df.set_index("time")

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

    nearest_GPS_station = mlcw_gps_distance_table.query("MLCW_Station==@select_MLCW_station")[
        "GPS_Stations"
    ].values[0]

    # search for the GPS fiole
    nearest_GPS_file = glob(os.path.join(gps_folder, nearest_GPS_station + "*"))[0]
    # get the vertical cumulative displacement from CSV file
    nearest_GPS_df = pd.read_csv(nearest_GPS_file, parse_dates=[0], index_col=[0], usecols=[0, 2])
    # resample the measurement to monthly
    nearest_GPS_monthly = nearest_GPS_df.resample("MS").mean()

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

    cache["MLCW_Station"].append(select_MLCW_station)
    cache["GPS_Station"].append(nearest_GPS_station)
    # - - - - - -
    mlcw_start = cdisp_mlcw_df.first_valid_index()
    gps_start = nearest_GPS_monthly.first_valid_index()
    cache["MLCW_Start"].append(mlcw_start)
    cache["GPS_Start"].append(gps_start)
    cache["Valid_Start"].append(max(mlcw_start, gps_start))
    # - - - - - -
    mlcw_end = cdisp_mlcw_df.last_valid_index()
    gps_end = nearest_GPS_monthly.last_valid_index()
    cache["MLCW_End"].append(mlcw_end)
    cache["GPS_End"].append(gps_end)
    cache["Valid_End"].append(min(mlcw_end, gps_end))
    # - - - - - -
```
