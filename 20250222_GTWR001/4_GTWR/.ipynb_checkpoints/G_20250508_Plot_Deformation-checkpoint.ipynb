```python
subsidence_areas = gpd.read_file(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\points_fld\support\sub_area_polygons.shp"
)
subsidence_areas.to_crs(crs="EPSG:3826", inplace=True)
subsidence_areas.plot(column="E_NAME")
```

```python
# ==============================================================================
# SECTION 1: USER CONFIGURATION
# ------------------------------------------------------------------------------
# HINT: This is the main control panel for the script.
#       Adjust the variables in this section to match your file paths and analysis needs.
# ==============================================================================

# --- Input File Configuration ---
# The main folder where your input data (the regression point files) are located.
TOP_FOLDER = "5_TestRun_105"

# The pattern used to find your regression point files within the TOP_FOLDER.
# The "*" is a wildcard. This example looks for all files ending in ".xz"
# that start with "BISQUARE".
REGPOINTS_FILE_PATTERN = "BISQUARE*.xz"

# --- Output Configuration ---
# The name of the main folder where all output figures will be saved.
FIGURE_OUTPUT_FOLDER = "figure_spatial_coeffs"


# ==============================================================================
# SECTION 2: SCRIPT EXECUTION
# ------------------------------------------------------------------------------
# The main logic of the program starts here.
# ==============================================================================

# --- Announce: Script Start ---
print("--- Starting Spatial-Temporal Coefficient Analysis ---")

# Discover all regression point files based on the configured pattern.
regpoints_files = glob(os.path.join(TOP_FOLDER, REGPOINTS_FILE_PATTERN))

if not regpoints_files:
    print(f"WARNING: No files found matching the pattern '{REGPOINTS_FILE_PATTERN}' in the folder '{TOP_FOLDER}'.")
    print("Script will now exit.")
    exit()

print(f"Found {len(regpoints_files)} regression point files to process.")
```

```python
# --- MAIN LOOP: Processing each regression point file ---
# This loop iterates through each of the main result files you want to analyze.
# HINT: To process all found files, change regpoints_files[:1] to just regpoints_files
for select_regpoints_file in tqdm(regpoints_files[:], desc="Processing Files"):
    
    # --- Announce: Current File ---
    print(f"\nProcessing file: {os.path.basename(select_regpoints_file)}")
    
    # Dynamically create a sub-folder for this specific file's outputs.
    # This keeps the results neatly organized.
    layer_from_name = "_".join(os.path.basename(select_regpoints_file).split("_")[1:3]).lower()
    figure_subfolder_path = os.path.join(TOP_FOLDER, FIGURE_OUTPUT_FOLDER, layer_from_name)
    
    # Create the output sub-directory if it doesn't exist.
    if not os.path.exists(figure_subfolder_path):
        os.makedirs(figure_subfolder_path)
        print(f"  - Created output sub-directory: {figure_subfolder_path}")

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    # Step 1: Load and Prepare Data
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    print("  - Loading and preparing regression point data...")
    
    # Load the full time-series data for all regression grid points.
    regpoints_df = pd.read_pickle(select_regpoints_file)
    
    # To perform a spatial search, we only need the unique locations of the grid points.
    # We create a "template" GeoDataFrame by filtering for a single time period.
    unique_regpoints = regpoints_df.query("time_period==1").copy()
    unique_regpoints = geospatial.convert_to_geodata(
        df=unique_regpoints, xcoord_col="X_TWD97", ycoord_col="Y_TWD97", crs_epsg="EPSG:3826"
    )

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    # Step 2: Pre-calculate which points fall into which districts.
    # This is an efficient "caching" step that avoids repeating the spatial query inside the loop.
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    print("  - Assigning grid points to administrative districts...")
    points_by_district = {}
    for idx, polygon in subsidence_areas.iterrows():
        district_name = polygon.E_NAME
        # Find all unique grid points that are within the current district polygon.
        regpoints_in_district = geospatial.find_points_within_polygon(
            points_gdf=unique_regpoints, polygon_geom=polygon.geometry
        )
        points_by_district[district_name] = regpoints_in_district
    
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    # Step 3: Loop through each district to generate plots.
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
    print(f"  - Generating plots for {len(points_by_district)} districts...")
    
    # HINT: To run for only one specific district for testing, you could change the loop to:
    # for select_district in ["TUKU"]:
    for select_district in tqdm(points_by_district.keys(), desc=f"Analyzing Districts for {layer_from_name}", position=1, leave=False):

        # Get the subset of grid points belonging to the current district.
        pointkey_byDistrict = points_by_district[select_district].index
        
        # If no points are in the district, skip to the next one.
        if len(pointkey_byDistrict) == 0:
            continue

        # Extract the full time-series data for just the points in this district.
        regpoints_byDistrict = regpoints_df.loc[pointkey_byDistrict, :].sort_values(by="time_period")
        
        # This dictionary will store the aggregated data for plotting the average trends.
        time_series_summary = {
            "time": [], "avg_coeffs": [], "coeffs_upper_std": [], "coeffs_lower_std": [],
            "avg_intercepts": [], "intercepts_upper_std": [], "intercepts_lower_std": []
        }
        
        # Create the plot figure and axes.
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 9), sharex=True)
        
        # --- Innermost Loop: Process each time period for the current district ---
        for t in regpoints_byDistrict["time_period"].unique():
            
            # Convert the numeric time period 't' into a proper datetime object for plotting on the x-axis.
            first_timepoint = pd.Timestamp(year=2016, month=5, day=1)
            current_time = first_timepoint + relativedelta(months=t)
            
            # Get all data for the current time step.
            regpoints_byTime = regpoints_byDistrict.query("time_period==@t")
            
            # --- Calculate Statistics for CUMDISP Coefficient ---
            coefficients_byTime = regpoints_byTime["CUMDISP_coeff"]
            average_coeff = coefficients_byTime.mean()
            stdev_coeff = coefficients_byTime.std()
            
            # --- Calculate Statistics for Intercept ---
            intercept_byTime = regpoints_byTime["Intercept"]
            average_intercept = intercept_byTime.mean()
            stdev_intercept = intercept_byTime.std()
            
            # --- Plotting: Individual Data Points (as a scatter cloud) ---
            # Plot all individual coefficient values at this time step as light grey points.
            ax1.plot([current_time] * len(coefficients_byTime), coefficients_byTime.values,
                     marker="o", linestyle=" ", markeredgecolor="lightgrey",
                     markerfacecolor="none", markeredgewidth=0.1, alpha=0.3)
            
            # Plot all individual intercept values at this time step.
            ax2.plot([current_time] * len(intercept_byTime), intercept_byTime.values,
                     marker="o", linestyle=" ", markeredgecolor="lightgrey",
                     markerfacecolor="none", markeredgewidth=0.1, alpha=0.3)

            # --- Storing Aggregated Data for Trend Lines ---
            time_series_summary["time"].append(current_time)
            time_series_summary["avg_coeffs"].append(average_coeff)
            time_series_summary["coeffs_upper_std"].append(average_coeff + stdev_coeff)
            time_series_summary["coeffs_lower_std"].append(average_coeff - stdev_coeff)
            
            time_series_summary["avg_intercepts"].append(average_intercept)
            time_series_summary["intercepts_upper_std"].append(average_intercept + stdev_intercept)
            time_series_summary["intercepts_lower_std"].append(average_intercept - stdev_intercept)

        # Convert the summary dictionary to a DataFrame for easy plotting and saving.
        summary_df = pd.DataFrame(time_series_summary).set_index("time")
        
        # --- Plotting: Average Trend and Standard Deviation Bands ---
        ax1.plot(summary_df.index, summary_df["avg_coeffs"], color="black", label="Average Coefficient (CUMDISP)", lw=1)
        ax1.fill_between(summary_df.index, summary_df["coeffs_lower_std"], summary_df["coeffs_upper_std"],
                         color="green", alpha=0.2, label="Std. Dev. Range")

        ax2.plot(summary_df.index, summary_df["avg_intercepts"], color="black", label="Average Intercept", lw=1)
        ax2.fill_between(summary_df.index, summary_df["intercepts_lower_std"], summary_df["intercepts_upper_std"],
                         color="blueviolet", alpha=0.2, label="Std. Dev. Range")
        
        # --- Final Plot Configuration and Saving ---
        ax1.set_ylabel("Coefficient (CUMDISP)")
        ax2.set_ylabel("Intercept")
        
        for ax in [ax1, ax2]:
            visualize.configure_axis(ax=ax, hide_spines=["top", "right"])
            visualize.configure_legend(ax=ax, fontsize_base=10)
            visualize.configure_datetime_ticks(ax=ax)
            ax.axhline(0, color='red', linestyle=':', linewidth=1) # Add a zero line for reference
            ax.text(0.02, 0.05, s=f"{len(pointkey_byDistrict)} grid points",
                    transform=ax.transAxes, fontsize=10, fontweight="bold")

        fig.suptitle(f"{select_district} District ({layer_from_name})", fontsize=14, fontweight='bold')
        fig.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make room for suptitle
        
        # Save both the plot image and the summary data.
        fig_outpath = os.path.join(figure_subfolder_path, f"{select_district}.png")
        excel_outpath = os.path.join(figure_subfolder_path, f"{select_district}_summary_data.xlsx")
        
        summary_df.to_excel(excel_outpath)
        visualize.save_figure_with_exact_dimensions(fig=fig, savepath=fig_outpath, width_px=2000, height_px=1120, dpi=300)
        
        plt.close(fig) # Close the figure to free memory.

# --- Announce: Script Completion ---
print("\n--- Analysis Complete ---")
print(f"All figures and summary data have been saved to sub-folders within: {os.path.join(TOP_FOLDER, FIGURE_OUTPUT_FOLDER)}")
```
