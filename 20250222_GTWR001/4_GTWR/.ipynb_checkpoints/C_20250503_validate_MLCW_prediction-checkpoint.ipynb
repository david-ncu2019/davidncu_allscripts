MLCW Ground Truth Validation Analysis

```python
# =========================================================================
# FILE PATHS
# =========================================================================
mlcw_measurements_path = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\3_MGTWR\2_Test_Run001\20250416_GWR_InputData_CUMDISP_MLCW_InSAR.csv"
regression_points_dir = r"4_Test_Run003"
regression_points_pattern = "GTWR_*.xz"
buffer_points_path = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\points_fld\support\grid_pnt_inBuffer2km.shp"
mlcw_stations_path = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\points_fld\mlcw_twd97.shp"

# =========================================================================
# PROCESSING CONFIGURATION
# =========================================================================
station_ids = ["TUKU"]  # Process multiple stations
buffer_radius = 600
time_sort_column = "monthly"
station_id_column = "STATION"

# =========================================================================
# PLOT CONFIGURATION
# =========================================================================
# primary_measurement_column = "Layer_1"
prediction_column = "MLCW_pred"
insar_measurement_column = "CUMDISP_measure"
show_insar_measurements = False
figure_width = 12
figure_height = 4
primary_color = "blue"
prediction_color = "lightgrey"
insar_color = "lime"
```

```python
# =========================================================================
# LOAD DATA
# =========================================================================

# Load MLCW measurement data
mlcw_measurements_df = pd.read_csv(mlcw_measurements_path)

# Load all regression point files
regression_point_files = glob(os.path.join(regression_points_dir, regression_points_pattern))

if not regression_point_files:
    raise ValueError(f"No regression files found matching pattern: {regression_points_pattern}")

regression_data_dict = {}

for file_path in regression_point_files:
    file_id = os.path.basename(file_path).replace(".xz", "")
    regression_data_dict[file_id] = pd.read_pickle(file_path)

# Load spatial data
buffer_points_gdf = gpd.read_file(buffer_points_path).set_index("PointKey")
mlcw_stations_gdf = gpd.read_file(mlcw_stations_path)

# =========================================================================
# VISUALIZE SPATIAL DISTRIBUTION
# =========================================================================
fig = plt.figure()
ax = fig.add_subplot(111)
buffer_points_gdf.plot(ax=ax)
mlcw_stations_gdf.plot(ax=ax)
plt.title("Station and Buffer Point Distribution")
plt.show()
```

```python
station_id = station_ids[0]

# Filter station measurements
station_measurements_df = mlcw_measurements_df.query(
    f"{station_id_column}==@station_id"
).sort_values(time_sort_column)

# if len(station_measurements_df) == 0:
#     print(f"Warning: No data found for station {station_id}")
#     continue

# Get station geometry
station_geometry_gdf = mlcw_stations_gdf.query(f"{station_id_column}==@station_id")

# if len(station_geometry_gdf) == 0:
#     print(f"Warning: Geometry not found for station {station_id}")
#     continue

# Find surrounding regression points
surrounding_points_gdf = geospatial.find_point_neighbors(
    central_point=station_geometry_gdf.iloc[0, :],
    target_points_gdf=buffer_points_gdf,
    central_key_column=station_id_column,
    buffer_radius=buffer_radius,
)

# =========================================================================
# PROCESS EACH REGRESSION POINT FILE
# =========================================================================

idx = 1
for regression_file_id, regression_df in regression_data_dict.items():

    # Extract coefficients for surrounding points
    surrounding_coeffs_df = regression_df.loc[surrounding_point_keys, :]

    # Create time-series plot
    fig = plt.figure(figsize=(figure_width, figure_height))
    ax = fig.add_subplot(111)

    # Plot predictions from each surrounding point
    for point_key in surrounding_point_keys:

        point_coeffs_df = surrounding_coeffs_df.loc[point_key, :].sort_values("time_period")
        mlcw_predictions = point_coeffs_df[prediction_column].values

        ax.plot(mlcw_predictions, color=prediction_color, marker="o", linestyle="--", alpha=0.2)

        # Optional InSAR measurements
        if show_insar_measurements and insar_measurement_column in point_coeffs_df.columns:
            insar_measurements = point_coeffs_df[insar_measurement_column].values
            ax.plot(
                insar_measurements, color=insar_color, marker="o", linestyle="--", alpha=0.2
            )

    # Plot actual MLCW measurements
    ax.plot(
        station_measurements_df[f"Layer_{idx}"].values,
        marker="o",
        linestyle="--",
        color=primary_color,
    )

    # Add plot labels
    plt.title(f"MLCW Validation: Station {station_id} - File {regression_file_id}")
    plt.xlabel("Time Period")
    plt.ylabel("Measurement Value")
    plt.tight_layout()
    plt.show()

    idx+=1
```
