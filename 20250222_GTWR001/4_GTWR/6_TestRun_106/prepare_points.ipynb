```python
folder = r"regpoints_filefld_addCUMDISP"
feather_files = glob(os.path.join(folder, "*.feather"))
feather_files[:5]
```

```python
tuku_points = pd.read_csv("TUKU_regpoints.csv")
tuku_pointkey = set(tuku_points["PointKey"])
```

```python
tuku_regpoints = pd.DataFrame(data=None, index=None)

# select_feather = feather_files[0]
for select_feather in tqdm(feather_files):
    regpoint_df = pd.read_feather(select_feather)
    regpoint_key = set(regpoint_df.index)
    mutual_points = regpoint_key.intersection(tuku_pointkey)
    if len(mutual_points) > 0:
        points_to_get = list(mutual_points)
        sub = regpoint_key.loc[points_to_get, :]
        tuku_regpoints = pd.concat([tuku_regpoints, sub])
```
