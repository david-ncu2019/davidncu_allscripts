```python
disp_mlcw_df = pd.read_pickle("MLCW_dataset/Monthly_MLCW_pca_DISPLACEMENT_v3.xz")
backtransform_disp_df = pd.read_pickle(
    "PostKriging_Backtransform_Data/BACKTRANSFORMED_Monthly_DISPLACEMENT_dU_CRFP_Full.xz"
)
```

```python
available_stations = disp_mlcw_df["STATION"].unique()
available_stations[:5]
```

```python
select_station = available_stations[0]

mlcw_disp_byStation = disp_mlcw_df.query("STATION==@select_station")

backtransform_disp_byStation = backtransform_disp_df.loc[select_station, measures_columns]

backtransform_disp_byStation.index = pd.to_datetime(
    [col[1:] for col in backtransform_disp_byStation.index]
)

mlcw_disp_byStation["DISPLACEMENT"] = mlcw_disp_byStation.index.map(backtransform_disp_byStation)
mlcw_disp_byStation = mlcw_disp_byStation.dropna(subset=["DISPLACEMENT"])

combined_input_df = pd.concat([combined_input_df, mlcw_disp_byStation])
```

```python
combined_input_df = pd.DataFrame(data=None, index=None)

measures_columns = [col for col in backtransform_disp_df.columns if col.startswith("N")]

# select_station = available_stations[0]
for select_station in tqdm(available_stations):
    
    mlcw_disp_byStation = disp_mlcw_df.query("STATION==@select_station")
    
    
    backtransform_disp_byStation = backtransform_disp_df.loc[select_station, measures_columns]
    
    backtransform_disp_byStation.index = pd.to_datetime(
        [col[1:] for col in backtransform_disp_byStation.index]
    )
    
    mlcw_disp_byStation["DISPLACEMENT"] = mlcw_disp_byStation.index.map(backtransform_disp_byStation)
    mlcw_disp_byStation = mlcw_disp_byStation.dropna(subset=["DISPLACEMENT"])

    combined_input_df = pd.concat([combined_input_df, mlcw_disp_byStation])

combined_input_df = combined_input_df.reset_index(drop=False)
combined_input_df = combined_input_df.sort_values(by=["time", "STATION"]).reset_index(drop=True)
```

```python
today_string = datetime.today().strftime("%Y%m%d")
combined_input_df.to_pickle(f"{today_string}_GWR_InputData_DISPLACEMENT_MLCW_InSAR.xz")
```
