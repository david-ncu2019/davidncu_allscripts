```python
def get_shapefiles(select_fld, fld_type="Validation", point_type="grid"):
    points_fld = [
        f for f in glob(os.path.join(select_fld, f"*{fld_type}*{point_type}*")) if os.path.isdir(f)
    ][0]
    points_shp = glob(os.path.join(points_fld, "*.shp"))
    return points_shp
```
