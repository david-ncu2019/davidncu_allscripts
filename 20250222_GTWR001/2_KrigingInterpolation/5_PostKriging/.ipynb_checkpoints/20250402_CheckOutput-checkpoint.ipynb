```python
mainfolder = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\4_Interpolation\001"
subfolders = [f for f in glob(mainfolder + "\\*") if os.path.isdir(f)]

# select_folder = subfolders[1]
expect_datetime = sorted([f"N{year}{str(month).zfill(2)}01" for month in range(1,13) for year in range(2014, 2022)])

for select_folder in subfolders:
    base = os.path.basename(select_folder)
    mlcw_shp = glob(os.path.join(select_folder, "*GA*Points*mlcw*", "*.shp"))
    mlcw_shp_datetime = [os.path.basename(f).split(".")[0].split("_")[-1] for f in mlcw_shp]
    datetime_missing = sorted(set(mlcw_shp_datetime).difference(set(expect_datetime)))
    print(base)
    print(datetime_missing)
```
