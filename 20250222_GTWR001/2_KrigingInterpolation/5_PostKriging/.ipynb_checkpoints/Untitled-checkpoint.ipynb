```python
# work with the first layer first

mlcw_df = pd.read_pickle("MLCW_dataset/Monthly_MLCW_CRFP_L1.xz")
mlcw_datetime_cols = [col for col in mlcw_df.columns if col.startswith("N")]
mlcw_datetime_cols[:5]
```

#### extract the mutual columns among all data sets

```python
from functools import reduce

all_pickles = glob("*.xz")

cache = [mlcw_datetime_cols, ]
for pkl_file in all_pickles:
    temp = pd.read_pickle(pkl_file)
    cols = [col for col in temp.columns if col.startswith("N")]
    cache.append(cols)
    
mutual_datetime_cols = sorted(reduce(set.intersection, map(set, cache)))
```

#### try with the first layer of MLCW

```python
extract_mlcw_df_byMutualDatetime = mlcw_df.loc[:, mutual_datetime_cols]
extract_mlcw_df_byMutualDatetime.columns = pd.to_datetime([col[1:] for col in mutual_datetime_cols])
extract_mlcw_df_byMutualDatetime = extract_mlcw_df_byMutualDatetime.interpolate(method='time', axis=1)
extract_mlcw_df_byMutualDatetime = extract_mlcw_df_byMutualDatetime.interpolate(method='bfill', axis=1)
extract_mlcw_df_byMutualDatetime.columns = mutual_datetime_cols
extract_mlcw_df_byMutualDatetime = extract_mlcw_df_byMutualDatetime.dropna(how='any', axis=1)
extract_mlcw_df_byMutualDatetime.head(5)
```

```python
mgtwr_ready_df = pd.DataFrame(None)
mlcw_cols = extract_mlcw_df_byMutualDatetime.columns
# select_col = mlcw_cols[0]

for select_col in tqdm(mlcw_cols[:]):
    prepare_df_byDatetime = pd.DataFrame(data=None, index=extract_mlcw_df_byMutualDatetime.index)
    prepare_df_byDatetime["X_TWD97"] = prepare_df_byDatetime.index.map(mlcw_df["X_TWD97"])
    prepare_df_byDatetime["Y_TWD97"] = prepare_df_byDatetime.index.map(mlcw_df["Y_TWD97"])
    
    mlcw_cols = extract_mlcw_df_byMutualDatetime.columns
    
    prepare_df_byDatetime.insert(loc=2, column="time", value=[pd.to_datetime(select_col[1:])]*len(prepare_df_byDatetime))
    
    prepare_df_byDatetime["MLCW_L1"] = prepare_df_byDatetime.index.map(extract_mlcw_df_byMutualDatetime[select_col])
    
    current_pkl = all_pickles[0]
    for current_pkl in all_pickles:
        
        predictor_colname = current_pkl.split(".")[0]
        predictor_df = pd.read_pickle(current_pkl)
        
        prepare_df_byDatetime[predictor_colname] = prepare_df_byDatetime.index.map(predictor_df[select_col])

    mgtwr_ready_df = pd.concat([mgtwr_ready_df, prepare_df_byDatetime], axis=0, ignore_index=True)

mgtwr_ready_df['time'] = mgtwr_ready_df['time'].dt.to_period('M')
mgtwr_ready_df['time'] = mgtwr_ready_df['time'].sub(mgtwr_ready_df['time'].iloc[0]).apply(lambda x: x.n)
```

```python
save_fld = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\6_Run_MGTWR"

mgtwr_ready_df.to_pickle(os.path.join(save_fld, "mgtwr_dset_MLCW_L1.xz"))
```
