```python
from appgeopy import *
from my_packages import *
```

```python
fpath = "gtwr_regpoints2_Prediction_Layer_1.csv"
df = pd.read_csv(fpath)
pointkey_arr = [f"X{int(x*1000)}Y{int(y*1000)}" for x, y in zip(df["X_TWD97"], df["Y_TWD97"])]
df.insert(loc=0, column="PointKey", value=pointkey_arr)
df.head()
```

```python
unique_time = df["pred_time"].unique()
unique_time[:3]
```

```python
# select_time = unique_time[0]
for select_time in tqdm(unique_time):
    
    subset = df.query("pred_time==@select_time").reset_index(drop=True)
    subset_geo = geospatial.convert_to_geodata(df=subset, xcoord_col="X_TWD97", ycoord_col="Y_TWD97", crs_epsg="EPSG:3826")
    
    new_cols = [name.replace(".", "_") if "." else name in name for name in subset_geo.columns]
    subset_geo.columns = new_cols
    subset_geo.to_file(f"output_shapefiles/gtwr_predict_time{str(select_time).zfill(3)}.shp")
```
