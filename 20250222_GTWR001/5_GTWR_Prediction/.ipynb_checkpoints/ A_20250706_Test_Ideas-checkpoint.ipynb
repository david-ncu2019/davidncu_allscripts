```python
# df = pd.read_csv(files[0])

for f, layer in zip(files[1:], [2,3,4]):
    df = pd.read_csv(f)
    df = df.drop(df.columns[0], axis=1)

    pointkey_arr = [
        f"X{int(x*1000)}Y{int(y*1000)}T{str(t).zfill(3)}"
        for x, y, t in zip(df["X_TWD97"], df["Y_TWD97"], df["monthly"])
    ]

    df.insert(loc=0, column="PointKey", value=pointkey_arr)

    df = df.set_index("PointKey")

    combined_df[f"Layer_{layer}"] = combined_df.index.map(df[f"Layer_{layer}"])

```
