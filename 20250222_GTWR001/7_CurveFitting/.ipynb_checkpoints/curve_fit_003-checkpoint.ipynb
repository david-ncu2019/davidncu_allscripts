```python
# Data loading and processing
base_directory = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\5_GTWR_Prediction\figure_timeseries_coeffs"

# Find all coefficient folders (excluding LAYER folders)
coefficient_folders = [
    folder
    for folder in os.listdir(base_directory)
    if "LAYER" not in folder
    and os.path.isdir(os.path.join(base_directory, folder))
]

for folder_name in tqdm(coefficient_folders[:1]):
    excel_files = glob(os.path.join(base_directory, folder_name, "*.xlsx"))

excel_files[:3]
```

```python
select_file = excel_files[2]

station_name = os.path.basename(select_file).split("_")[0]
# select_file = r"D:\\1000_SCRIPTS\\003_Project002\\20250222_GTWR001\\5_GTWR_Prediction\\figure_timeseries_coeffs\\Layer_1\\BAOZHONG_summary_data.xlsx"

data_frame = pd.read_excel(select_file, parse_dates=[0])
layer_name = os.path.basename(select_file).split(".")[0]


original_start_date = data_frame["time"].iloc[0]

# Create prediction time range from data start to end of 2024
prediction_timerange = pd.date_range(
    start=original_start_date, end=datetime(2024, 12, 31), freq="1MS"
)

# Convert time to cumulative days for modeling
time_days = (
    data_frame["time"].diff().apply(lambda x: x.days).fillna(0).cumsum().values
)

prediction_days = (
    pd.Series(prediction_timerange)
    .diff()
    .apply(lambda x: x.days)
    .fillna(0)
    .cumsum()
    .values
)

prediction_years = (
    prediction_timerange - original_start_date
).total_seconds() / (3600 * 24 * 365.25)

# Extract coefficient values for modeling
coefficient_values = data_frame["avg_coeffs"].values

data_frame = data_frame.set_index("time")

# Model fitting with exponential variogram
model_function = exponential

# Set up fitting parameters and bounds
parameter_bounds = (
    [np.min(time_days), np.min(coefficient_values), np.min(coefficient_values)],
    [np.max(time_days), np.max(coefficient_values), np.max(coefficient_values)],
)
initial_guess = [
    np.median(time_days),
    np.median(coefficient_values),
    coefficient_values[0],
]

# Fit model on valid (non-NaN) data points
valid_data_mask = ~np.isnan(coefficient_values)

exponential_fitted_params, exponential_parameter_covariance = curve_fit(
    model_function,
    time_days[valid_data_mask],
    coefficient_values[valid_data_mask],
    p0=initial_guess,
    bounds=parameter_bounds,
)

# Generate pure exponential predictions for entire timerange
exponential_predictions = model_function(
    prediction_days, *exponential_fitted_params
)

# Organize predictions with time index
prediction_dataframe = pd.DataFrame(
    data={"exponential": exponential_predictions}, index=prediction_timerange
)

prediction_dataframe["original"] = prediction_dataframe.index.map(
    data_frame["avg_coeffs"]
)

subset = prediction_dataframe.dropna(how="any")
subset["residual"] = subset["original"] - subset["exponential"]


DAMPED_NUM_COMPONENTS = 3
fitted_params, fitted_signal, metrics, t_years = fit_multi_damped_sine_wave(
    subset,
    num_components=DAMPED_NUM_COMPONENTS,
    value_column="residual",  # Use the new parameter here
)

sinwave_predictions = multi_damped_sine_model(
    prediction_years, fitted_params, DAMPED_NUM_COMPONENTS
)
sinwave_prediction_series = pd.Series(
    sinwave_predictions, index=prediction_timerange
)
prediction_dataframe["sinusoidal"] = prediction_dataframe.index.map(
    sinwave_prediction_series
)

prediction_dataframe["model"] = (
    prediction_dataframe["exponential"] + prediction_dataframe["sinusoidal"]
)
```
