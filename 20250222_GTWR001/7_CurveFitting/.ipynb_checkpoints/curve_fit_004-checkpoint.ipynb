```python
from appgeopy import *
from my_packages import *
from scipy.optimize import curve_fit, differential_evolution
from scipy.signal import periodogram
from skgstat.models import exponential
from statsmodels.tsa.seasonal import MSTL

# Set a professional plot style
plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["figure.figsize"] = (10, 3)
```

```python
mlcw_gdf = gpd.read_file(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\points_fld\mlcw_twd97.shp"
)
```

```python
# Define the folder path where model prediction results are stored.
model_folder = (
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\5_GTWR_Prediction"
)

# Define the specific kernel name used in the model to find relevant result files.
kernel_name = "bisquare"

# Analysis settings
START_DATE = pd.Timestamp(year=2016, month=5, day=1)

# Use `glob` to find all files within `model_folder` that contain the kernel name in their filename.
output_files = glob(os.path.join(model_folder, f"*{kernel_name}*.csv"))
# Display the list of found files.
output_files
```

```python
select_file = output_files[0]


df = pd.read_csv(select_file)

pointkey_arr = [
    f"X{int(x*1000)}Y{int(y*1000)}"
    for x, y in zip(df["X_TWD97"], df["Y_TWD97"])
]
# Insert this new 'PointKey' column at the beginning of the DataFrame.
df.insert(loc=0, column="PointKey", value=pointkey_arr)
# Set the 'PointKey' column as the DataFrame's index for efficient data lookup.
df = df.set_index("PointKey")

# Get a list of all unique point keys (i.e., all unique locations) in the dataset.
unique_pointkey = df.index.unique()
```

```python
select_pointkey = unique_pointkey[12]
mlcw_station = mlcw_gdf.query("PointKey==@select_pointkey").STATION.values[0]
print(mlcw_station)
df_byPointKey = df.loc[select_pointkey]

time_arr = pd.to_datetime(
    [
        START_DATE + relativedelta(months=int(time_period))
        for time_period in df_byPointKey["time_stamp"]
    ]
)

df_byPointKey["datetime"] = time_arr

df_byPointKey = df_byPointKey.set_index("datetime")

df_byPointKey.sort_index(inplace=True)

coefficient_values = df_byPointKey["CUMDISP"]
coefficient_values = coefficient_values.asfreq("MS")
coefficient_values.plot(marker="o", markerfacecolor="none")
```

```python
# Convert time to cumulative days for modeling
time_days = (
    pd.Series(time_arr).diff().apply(lambda x: x.days).fillna(0).cumsum().values
)
# Model fitting with exponential variogram
model_function = exponential

# Set up fitting parameters and bounds
parameter_bounds = (
    [np.min(time_days), np.min(coefficient_values), np.min(coefficient_values)],
    [np.max(time_days), np.max(coefficient_values), np.max(coefficient_values)],
)
initial_guess = [
    np.median(time_days),
    np.median(coefficient_values),
    coefficient_values[0],
]

# Fit model on valid (non-NaN) data points
valid_data_mask = ~np.isnan(coefficient_values)

exponential_fitted_params, exponential_parameter_covariance = curve_fit(
    model_function,
    time_days[valid_data_mask],
    coefficient_values[valid_data_mask],
    p0=initial_guess,
    bounds=parameter_bounds,
)

# Generate pure exponential predictions for entire timerange
exponential_predictions = pd.Series(data=model_function(
    time_days, *exponential_fitted_params
), index=df_byPointKey.index)
```

```python
plt.plot(coefficient_values)
plt.plot(exponential_predictions)
```

```python
seasonal_components = coefficient_values-exponential_predictions
seasonal_components[:5]
```

```python
# Detrend the series to make cycles more visible in the periodogram
detrended_series = seasonal_components - seasonal_components.rolling(window=12, center=True).mean().bfill().ffill()
seasonal_components.plot(label='original')
detrended_series.plot(label='detrend')
plt.legend()
```

```python
# A periodogram shows the "power" of different frequencies.
# fs=12 means our data is sampled 12 times per year (monthly).
frequencies, power = periodogram(detrended_series.dropna(), fs=12)

# We convert frequency (cycles/year) to period (months/cycle)
# Period = 1 / (Frequency / Sampling Rate) = 12 / Frequency
# We ignore frequency 0, which is the trend/mean.
periods = 12 / frequencies[1:]
power = power[1:]

# Create the plot
plt.figure(figsize=(14, 6))
plt.plot(periods, power)
plt.title('Periodogram: Power vs. Period')
plt.xlabel('Period (in Months)')
plt.ylabel('Power (Strength of Cycle)')
plt.xscale('log') # Use a log scale to see both short and long periods clearly
plt.grid(True, which="both", ls="--")
```

```python
# Find and annotate the strongest peaks
peak_indices = np.argsort(power)[-5:] # Get indices of 5 strongest peaks
for i in peak_indices:
    peak_period = periods[i]
    peak_power = power[i]
    # Only annotate significant peaks to avoid clutter
    if peak_power > np.median(power) * 2:
        plt.axvline(x=peak_period, color='red', linestyle='--', alpha=0.7)
        plt.text(peak_period, peak_power, f' {peak_period:.1f} months', color='red')

plt.show()
```

```python
if not significant_periods:
    print("\nNo significant periods were found. Cannot perform MSTL decomposition.")
    print("The data may be pure trend or random noise.")
else:
    print(f"\nPerforming MSTL decomposition with discovered periods: {significant_periods}...")
    
# Use the discovered periods in MSTL
mstl = MSTL(seasonal_components, periods=significant_periods)
results = mstl.fit()

print("Decomposition complete. Plotting results...")
plt.rcParams["figure.figsize"] = (10, 10)
# The MSTL results object has a convenient built-in plot method
results.plot()
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Create a DataFrame to hold all our Fourier terms
fourier_terms = pd.DataFrame(index=seasonal_components.index)

# We will use 1 harmonic (one sine/cosine pair) for each period.
harmonics = 1 

# Loop through each period and create the corresponding sine/cosine terms
for p in significant_periods:
    for i in range(1, harmonics + 1):
        # Sine term
        fourier_terms[f'sin_{p}_{i}'] = np.sin(2 * np.pi * i * np.arange(len(seasonal_components)) / p)
        # Cosine term
        fourier_terms[f'cos_{p}_{i}'] = np.cos(2 * np.pi * i * np.arange(len(seasonal_components)) / p)

print(fourier_terms.head())
```

```python
sarimax_model = SARIMAX(
    endog=seasonal_components,
    exog=fourier_terms,
    order=(1, 1, 1),
    seasonal_order=(0, 0, 0, 0) # No internal seasonality needed
)

# Fit the model
print("Fitting the SARIMAX model...")
sarimax_results = sarimax_model.fit(disp=False)
print("Model fitting complete.")

# Print the summary to check coefficients. We expect many of the
# sin/cos terms to be statistically significant (P>|z| < 0.05).
print(sarimax_results.summary())
```

```python
simple_model = SARIMAX(
    seasonal_components,
    order=(0, 1, 1), # p=0, d=1, q=1
    seasonal_order=(0, 0, 0, 0)
)

# 2. Fit the new model
print("Fitting the simpler model...")
simple_results = simple_model.fit(disp=False)

# 3. Print the new summary and compare
print("\n--- Summary of the New, Simpler Model ---")
print(simple_results.summary())
```

```python
forecast_horizon = 12
print(f"\nGenerating forecast for the next {forecast_horizon} months...")

forecast_object = simple_results.get_forecast(steps=forecast_horizon)

# Extract the mean forecast and confidence intervals
forecast_mean = forecast_object.predicted_mean
conf_int = forecast_object.conf_int(alpha=0.1)


# --- 4. VISUALIZE THE FINAL FORECAST ---
plt.figure(figsize=(16, 8))
plt.plot(seasonal_components, label='Historical Data')
plt.plot(forecast_mean, label='Final Forecast (ARIMA(0,1,1))', color='green', linestyle='--')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='green', alpha=0.2, label='95% Confidence Interval')

plt.title('Final Forecast using Optimal ARIMA(0,1,1) Model', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
