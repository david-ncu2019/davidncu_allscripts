```python
# import pandas as pd
# import geopandas as gpd
# import numpy as np
# import matplotlib.pyplot as plt
from pathlib import Path

# from dateutil.relativedelta import relativedelta
from typing import Tuple

from appgeopy import *
from my_packages import *
from scipy.optimize import curve_fit
from skgstat.models import exponential
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# =============================================================================
# 1. CONFIGURATION
# =============================================================================
class Config:
    """A centralized class for all user-configurable parameters."""

    # --- Input File Paths ---
    SHAPEFILE_PATH = Path(
        r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\points_fld\mlcw_twd97.shp"
    )
    MODEL_FOLDER = Path(
        r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\5_GTWR_Prediction"
    )

    # --- Output Folder ---
    OUTPUT_FOLDER = Path(r"Curve_Fitting_Results")

    # --- Analysis Settings ---
    KERNEL_NAME: str = "bisquare"
    START_DATE = pd.Timestamp(year=2016, month=5, day=1)

    # --- Plotting Style ---
    plt.style.use("seaborn-v0_8-talk")
    plt.rcParams["figure.figsize"] = (12, 5)
```

```python
# =============================================================================
# 2. HELPER FUNCTIONS
# =============================================================================


def load_and_prepare_data(file_path: Path) -> pd.DataFrame:
    """Loads CSV data, creates a 'PointKey', and sets it as the index."""
    df = pd.read_csv(file_path)
    pointkey_arr = [
        f"X{int(x*1000)}Y{int(y*1000)}"
        for x, y in zip(df["X_TWD97"], df["Y_TWD97"])
    ]
    df.insert(loc=0, column="PointKey", value=pointkey_arr)
    df.set_index("PointKey", inplace=True)
    return df


def get_time_series_for_point(
    df: pd.DataFrame, point_key: str, start_date: pd.Timestamp
) -> Tuple[pd.Series, np.ndarray]:
    """Extracts and prepares the time series and cumulative days for a single point."""
    df_point = df.loc[point_key].copy()
    time_arr = [
        start_date + relativedelta(months=int(t))
        for t in df_point["time_stamp"]
    ]
    df_point["datetime"] = time_arr
    df_point = df_point.set_index("datetime").sort_index()
    coefficient_series = df_point["CUMDISP"].asfreq("MS")
    time_days = (
        pd.Series(coefficient_series.index)
        .diff()
        .apply(lambda x: x.days)
        .fillna(0)
        .cumsum()
        .values
    )
    return coefficient_series, time_days


def fit_exponential_model(
    series: pd.Series, time_days: np.ndarray
) -> Tuple[np.ndarray, pd.Series]:
    """Fits an exponential model using the original, correct logic."""
    valid_mask = ~np.isnan(series)
    y_data = series[valid_mask]
    x_data = time_days[valid_mask]

    if len(y_data) < 3:
        raise ValueError("Not enough valid data points to fit model.")

    parameter_bounds = (
        [np.min(x_data), np.min(y_data), np.min(y_data)],
        [np.max(x_data), np.max(y_data), np.max(y_data)],
    )
    initial_guess = [np.median(x_data), np.median(y_data), y_data.iloc[0]]

    params, _ = curve_fit(
        exponential, x_data, y_data, p0=initial_guess, bounds=parameter_bounds
    )
    predictions = pd.Series(
        data=exponential(time_days, *params), index=series.index
    )
    return params, predictions


def calculate_model_metrics(
    observation: pd.Series, prediction: pd.Series
) -> dict:
    denominator = np.sum(observation)
    if denominator == 0:
        pbias = np.nan
    else:
        pbias = 100.0 * np.sum(prediction - observation) / denominator

    metrics = {
        "r_square": r2_score(observation, prediction),
        "rmse": np.sqrt(mean_squared_error(observation, prediction)),
        "mae": mean_absolute_error(observation, prediction),
        "pbias": pbias,
    }
    return metrics
```

```python
Config.OUTPUT_FOLDER.mkdir(parents=True, exist_ok=True)
print(f"Output figures will be saved to: {Config.OUTPUT_FOLDER}")

output_files = sorted(
    list(Config.MODEL_FOLDER.glob(f"*{Config.KERNEL_NAME}*.csv"))
)
if not output_files:
    raise FileNotFoundError(
        f"No files with kernel '{Config.KERNEL_NAME}' found in {Config.MODEL_FOLDER}"
    )

selected_file = output_files[0]
df_processed = load_and_prepare_data(selected_file)
mlcw_gdf = gpd.read_file(Config.SHAPEFILE_PATH)

unique_point_keys = sorted(df_processed.index.unique())
total_points = len(unique_point_keys)
print(f"Found {total_points} unique points to process.\n")
```

```python
i = 4
point_key = unique_point_keys[i]

station_name = "Unknown"  # Default name in case of error
try:
    station_name = mlcw_gdf.query("PointKey == @point_key").STATION.values[0]
    print(
        f"[{i+1}/{total_points}] Processing Station: {station_name} ({point_key})"
    )

    coefficient_series, time_days = get_time_series_for_point(
        df_processed, point_key, Config.START_DATE
    )

    fitted_params, predicted_values = fit_exponential_model(
        coefficient_series, time_days
    )

except Exception as e:
    print(f"  -> FAILED for station {station_name} ({point_key}). Reason: {e}")

coefficient_series.plot()
predicted_values.plot()
```

```python
from sklearn.model_selection import TimeSeriesSplit

# ========================================================================
# STEP 1: DATA PREPROCESSING AND VALIDATION
# ========================================================================

# Create the initial DataFrame
temp = pd.DataFrame(
    {"coefficients": coefficient_series, "exp_trend": predicted_values}
)
temp["detrend"] = temp["coefficients"] - temp["exp_trend"]
temp["ma_detrend"] = smoothing.simple_moving_average(temp["detrend"], window_size=13)

data = temp[["ma_detrend"]]
data_days_interval = convert_to_fulltime(data)
seasonality_full = analysis.find_seasonality(data_days_interval, "ma_detrend")

# Prepare data for modeling
data_days_interval["dummy"] = np.arange(len(data_days_interval))
data_clean = data_days_interval.dropna(how="any")

time_values = data_clean["dummy"].values
observed_values = data_clean["ma_detrend"].values

# Cross-validation setup
tscv = TimeSeriesSplit(n_splits=5, test_size=len(observed_values)//8)
```

```python
# ========================================================================
# STEP 3: USE CROSS-VALIDATION TO SELECT OPTIMAL NUMBER OF COMPONENTS
# ========================================================================

# Test different numbers of components to find optimal
# component_options = [3, 5, 7, 10, 12]  # Different numbers to test
component_options = np.arange(5, 51, 5)  # Different numbers to test
best_performance = float('inf')
best_n_components = None
best_seasonality = None
all_cv_results = {}

for n_comp in component_options:
    # Select top N components for this test
    seasonality_test = seasonality_full.nlargest(n_comp, "Amplitude")
    
    # Filter by reasonable periods
    seasonality_test = seasonality_test[
        (seasonality_test["Period (days)"] >= 30) & 
        (seasonality_test["Period (days)"] <= 2000)
    ]
    
    if len(seasonality_test) == 0:
        continue
        
    # Extract parameters for this component set
    amplitudes = seasonality_test["Amplitude"].values
    periods = seasonality_test["Period (days)"].values
    phase_shifts = seasonality_test["Phase"].values
    baseline = np.mean(observed_values)
    
    # Cross-validation for this component set
    cv_scores = []
    
    for fold, (train_idx, test_idx) in enumerate(tscv.split(time_values)):
        train_time = time_values[train_idx]
        train_obs = observed_values[train_idx]
        test_time = time_values[test_idx]
        test_obs = observed_values[test_idx]
        
        try:
            fitted_signal_train = modeling.fit_sinusoidal_model(
                time_values=train_time,
                observed_values=train_obs,
                amplitudes=amplitudes,
                periods=periods,
                phase_shifts=phase_shifts,
                baseline=baseline,
                predict_time=test_time,
            )
            
            if fitted_signal_train is not None:
                mae = mean_absolute_error(test_obs, fitted_signal_train)
                rmse = np.sqrt(mean_squared_error(test_obs, fitted_signal_train))
                cv_scores.append({"fold": fold, "mae": mae, "rmse": rmse})
                
        except Exception as e:
            continue
    
    # Calculate average performance for this component set
    if cv_scores:
        avg_mae = np.mean([score["mae"] for score in cv_scores])
        avg_rmse = np.mean([score["rmse"] for score in cv_scores])
        std_mae = np.std([score["mae"] for score in cv_scores])
        
        all_cv_results[n_comp] = {
            "avg_mae": avg_mae,
            "avg_rmse": avg_rmse,
            "std_mae": std_mae,
            "cv_scores": cv_scores,
            "seasonality": seasonality_test
        }
        
        # Select best performing component set
        if avg_mae < best_performance:
            best_performance = avg_mae
            best_n_components = n_comp
            best_seasonality = seasonality_test

if best_seasonality is None:
    raise ValueError("No successful model configurations found!")
```

```python
# ========================================================================
# STEP 4: FIT FINAL MODEL USING CV-SELECTED COMPONENTS
# ========================================================================

# Now use the CV-selected best components for final fitting
amplitudes = best_seasonality["Amplitude"].values
periods = best_seasonality["Period (days)"].values
phase_shifts = best_seasonality["Phase"].values
baseline = np.mean(observed_values)

final_fitted = modeling.fit_sinusoidal_model(
    time_values=time_values,
    observed_values=observed_values,
    amplitudes=amplitudes,
    periods=periods,
    phase_shifts=phase_shifts,
    baseline=baseline
)
```

```python
# ========================================================================
# STEP 5: PREDICTION WITH CV-INFORMED UNCERTAINTY
# ========================================================================
    
start_date = temp.first_valid_index()
end_date = temp.last_valid_index()
prediction_end = "2024-12-31"

# Create monthly prediction timeline
historical_end = pd.to_datetime(end_date)
prediction_start = historical_end + pd.DateOffset(months=0)
prediction_timeline = pd.date_range(start=prediction_start, end=prediction_end, freq='MS')

# Calculate prediction indices
historical_start = data_clean.index.min()
prediction_indices = [(date - historical_start).days for date in prediction_timeline]
prediction_indices = np.array(prediction_indices)

# Generate predictions
predictions = modeling.fit_sinusoidal_model(
    time_values=time_values,
    observed_values=observed_values,
    amplitudes=amplitudes,
    periods=periods,
    phase_shifts=phase_shifts,
    baseline=baseline,
    predict_time=prediction_indices
)

# Use CV performance to estimate prediction uncertainty
# This is more realistic than just using residuals
cv_performance = all_cv_results[best_n_components]
prediction_uncertainty = cv_performance["avg_mae"]  # Use CV MAE as uncertainty estimate

# Create confidence intervals based on CV performance
confidence_levels = [0.68, 0.95]
prediction_intervals = {}

for conf_level in confidence_levels:
    z_score = 1.0 if conf_level == 0.68 else 1.96
    margin = z_score * prediction_uncertainty
    prediction_intervals[f"{int(conf_level*100)}%"] = {
        "lower": predictions - margin,
        "upper": predictions + margin
    }

# Model evaluation
final_metrics = analysis.evaluate_model_fit(
    pd.Series(observed_values), 
    pd.Series(final_fitted)
)
```

```python
# ========================================================================
# STEP 6: MODEL EVALUATION AND DIAGNOSTICS
# ========================================================================

# Calculate comprehensive model performance metrics
final_metrics = analysis.evaluate_model_fit(
    pd.Series(observed_values), pd.Series(final_fitted)
)

final_metrics
```

```python
# ========================================================================
# STEP 7: RESULTS COMPILATION
# ========================================================================

# ========================================================================
# RESULTS WITH CV-INFORMED SELECTIONS
# ========================================================================

results = {
    "data_info": {
        "start_date": start_date,
        "end_date": end_date,
        "n_observations": len(observed_values)
    },
    "cv_selection_process": {
        "tested_components": component_options,
        "all_cv_results": all_cv_results,
        "selected_n_components": best_n_components,
        "selection_criterion": "lowest_cv_mae"
    },
    "model_details": {
        "seasonality_components": best_seasonality,
        "n_components": len(best_seasonality),
        "cv_performance": cv_performance
    },
    "model_fit": {
        "fitted_values": final_fitted,
        "observed_values": observed_values,
        "residuals": observed_values - final_fitted,
        "metrics": final_metrics
    },
    "predictions": {
        "timeline": prediction_timeline,
        "values": predictions,
        "confidence_intervals": prediction_intervals,
        "uncertainty_source": "cross_validation_performance"
    }
}
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
title = "Seasonal Time Series Model"

# Extract data for plotting
# Create proper monthly timeline for historical data
historical_timeline = pd.date_range(
    start=results['data_info']['start_date'],
    end=results['data_info']['end_date'],
    freq='MS'
)
observed_values = results["model_fit"]["observed_values"]
fitted_values = results["model_fit"]["fitted_values"]

prediction_timeline = results["predictions"]["timeline"]
predictions = results["predictions"]["values"]
lower_68 = results["predictions"]["confidence_intervals"]["68%"]["lower"]
upper_68 = results["predictions"]["confidence_intervals"]["68%"]["upper"]
lower_95 = results["predictions"]["confidence_intervals"]["95%"]["lower"]
upper_95 = results["predictions"]["confidence_intervals"]["95%"]["upper"]

# Plot historical data
ax.plot(
    historical_timeline,
    temp["detrend"].values,
    "o-",
    label="Original Data",
    color="black",
    alpha=0.7,
    markersize=4,
)
ax.plot(
    historical_timeline,
    observed_values,
    "o-",
    label="Smoothed Data",
    color="blue",
    alpha=0.7,
    markersize=4,
)
ax.plot(
    historical_timeline,
    fitted_values,
    "-",
    label="Fitted Model",
    color="red",
    linewidth=2,
)

# Plot predictions
ax.plot(
    prediction_timeline,
    predictions,
    "-",
    label="Predictions",
    color="green",
    linewidth=2,
)

# Add confidence intervals as filled areas
ax.fill_between(
    prediction_timeline,
    lower_95,
    upper_95,
    alpha=0.2,
    color="green",
    label="95% Confidence",
)
ax.fill_between(
    prediction_timeline,
    lower_68,
    upper_68,
    alpha=0.3,
    color="green",
    label="68% Confidence",
)

# Add vertical line to separate historical and prediction periods
ax.axvline(x=historical_timeline[-1], color="gray", linestyle="--", alpha=0.7)

# Formatting
ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.set_title(title)
ax.legend()
ax.grid(True, alpha=0.3)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
