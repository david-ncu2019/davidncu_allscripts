## Merge and Export to Shapefiles

```python
prediction_output_folder = "3__PredictionOutput"
layer = "Layer_4"
output_files = glob(os.path.join(prediction_output_folder, layer, "*.feather"))
output_files[:3]
```

```python
combined_output_df = pd.DataFrame(data=None, index=None)

# select_file = output_files[0]
for select_file in tqdm(output_files[:]):
    df = pd.read_feather(select_file)
    new_cols = [
        col.replace(".", "_") if "." in col else col for col in df.columns
    ]
    pointkey_arr = [
        f"X{int(x*1000)}Y{int(y*1000)}"
        for x, y in zip(df["X_TWD97"], df["Y_TWD97"])
    ]
    df.columns = new_cols
    df.insert(loc=0, column="PointKey", value=pointkey_arr)
    combined_output_df = pd.concat([combined_output_df, df], ignore_index=True)

show(combined_output_df)
```

```python
# save combined dataframe to feather for later use

combined_output_df.to_feather(
    os.path.join(prediction_output_folder, f"{layer}.feather")
)
```
