```python
# load points that stay within buffer of MLCW #### REMEMBER TO RUN THIS IN ADVANCE

# points_within_buffer = pd.read_pickle(
#     r"CUMDISP_within_MLCW_buffer100m_saveqgis_Oct2025.xz"
# ) # old version

points_within_buffer = pd.read_pickle(
    r"CUMDISP_within_MLCW_buffer100m_saveqgis_Oct2025_v2.xz"
)

# load timeseries of cumulative displacement of all points

#### old version
# cumdisp_df = pd.read_pickle(r"Monthly_CUMDISP_saveqgis_Oct2025.xz")
# cumdisp_df = cumdisp_df.set_index("PointKey")

#### new version
cumdisp_df = pd.read_pickle(
    r"D:\1000_SCRIPTS\003_Project002\20250917_GTWR002\1_PrepareDatasets\LevelingBechmark\CORRECTED_Monthly_CUMDISP_CRFP_saveqgis_Oct2025.xz"
)
show(cumdisp_df)
```

```python
def change_colname(colname):
    # Remove the leading 'N'
    date_str = colname[1:]

    # Convert to datetime object
    date_obj = datetime.strptime(date_str, "%Y%m%d")

    # Add one day
    date_obj += timedelta(days=1)

    # Convert back to string with 'N' prefix and formatted date
    output_str = "N" + date_obj.strftime("%Y%m%d")

    return output_str
```

```python
combine_df = pd.DataFrame(data=None, index=None)

unique_stations = points_within_buffer["STATION"].unique()

# select_station = "TUKU"
for select_station in unique_stations[:]:

    pointkey_byStation = points_within_buffer.query("STATION==@select_station")[
        "PointKey"
    ].tolist()

    cumdisp_byStation = cumdisp_df.loc[pointkey_byStation]
    measure_cols = [
        col for col in cumdisp_byStation.columns if col.startswith("D")
    ]
    cumdisp_byStation = cumdisp_byStation.loc[:, measure_cols]
    average_cumdisp_byStation = cumdisp_byStation.mean(axis=0)
    average_cumdisp_byStation_df = pd.DataFrame(
        average_cumdisp_byStation, columns=[select_station]
    )

    combine_df = pd.concat([combine_df, average_cumdisp_byStation_df.T], axis=0)

combine_df.columns = list(map(change_colname, combine_df.columns.tolist()))
combine_df.head(5)
```

```python
# we will borrow the old file to get the information

temp = pd.read_pickle(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\5_PostKriging\PostKriging_Backtransform_Data\mlcw_points\BACKTRANSFORMED_Monthly_DISPLACEMENT_dU_CRFP_Full.xz"
)

temp = temp.iloc[:, :3]
for col in tqdm(combine_df.columns):
    temp[col] = temp.index.map(combine_df[col])

show(temp)
```

# 2

After we get that table, run the following code snippet to get this table:



```python
cumdisp_mlcw_df = pd.read_pickle(
    r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\2_KrigingInterpolation\5_PostKriging\MLCW_dataset\Monthly_MLCW_pca_CUMDISP_v2.xz"
)
show(cumdisp_mlcw_df)
```

```python
temp = pd.read_pickle(r"AVERAGED_Monthly_CUMDISP_mlcw_saveqgis_Oct2025_corrected.xz")
backtransform_cumdisp_df = temp.copy()
cumdisp_columns = [
    col for col in backtransform_cumdisp_df.columns if col.startswith("N")
]
show(backtransform_cumdisp_df)
```

### calibration points

```python
# available_stations = cumdisp_mlcw_df["STATION"].unique()
available_stations = list(temp.index)

combined_input_df = pd.DataFrame(data=None, index=None)

# select_station = available_stations[0]
for select_station in tqdm(available_stations):
    # for select_station in ["TUKU"]:

    mlcw_cumpdisp_byStation = cumdisp_mlcw_df.query("STATION==@select_station")

    backtransform_cumdisp_byStation = backtransform_cumdisp_df.loc[
        select_station, cumdisp_columns
    ]

    dates = pd.to_datetime(
        [col[1:] for col in backtransform_cumdisp_byStation.index]
    )
    backtransform_cumdisp_byStation.index = dates

    mlcw_cumpdisp_byStation["CUMDISP"] = mlcw_cumpdisp_byStation.index.map(
        backtransform_cumdisp_byStation
    )

    mlcw_cumpdisp_byStation = mlcw_cumpdisp_byStation.loc["2016-4":"2021"]
    mlcw_cumpdisp_byStation = mlcw_cumpdisp_byStation.fillna(value=0)
    # mlcw_cumpdisp_byStation = mlcw_cumpdisp_byStation.dropna(subset=["CUMDISP"])

    combined_input_df = pd.concat([combined_input_df, mlcw_cumpdisp_byStation])

combined_input_df = combined_input_df.reset_index(drop=False)
combined_input_df = combined_input_df.sort_values(
    by=["time", "STATION"]
).reset_index(drop=True)

show(combined_input_df)
```

```python
today_string = datetime.today().strftime("%Y%m%d")
combined_input_df.to_pickle(f"{today_string}_GTWR_InputData_MLCW_InSAR.xz")
```
