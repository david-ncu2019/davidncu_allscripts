```python
folder_path = r"D:\1000_SCRIPTS\003_Project002\20250222_GTWR001\5_GTWR_Prediction\3__PredictionOutput\Layer_4"
shp_files = glob(os.path.join(folder_path, "*.shp"))
shp_files[:3]
```

```python
combined_output = pd.DataFrame(data=None, index=None)

# select_shp = shp_files[0]
for select_shp in tqdm(shp_files):

    gdf = gpd.read_file(select_shp, read_geometry=False, engine="pyogrio")
    
    pointkey_arr = [f"X{int(x*1000)}Y{int(y*1000)}" for x, y in zip(gdf["POINT_X"], gdf["POINT_Y"])]
    
    gdf = gdf.rename({"POINT_X":"X_TWD97", "POINT_Y":"Y_TWD97"}, axis=1)
    gdf.insert(loc=0, column="PointKey", value=pointkey_arr)
    
    combined_output = pd.concat([combined_output, gdf])

combined_output
```
