## Merge and Export to Shapefiles

```python
prediction_output_folder = "3__PredictionOutput"
layer = "Layer_2"
output_files = glob(os.path.join(prediction_output_folder, layer, "*.feather"))
output_files[:3]
```

```python
combined_output_df = pd.DataFrame(data=None, index=None)

# select_file = output_files[0]
for select_file in tqdm(output_files[:]):
    df = pd.read_feather(select_file)
    new_cols = [col.replace(".", "_") if "." in col else col for col in df.columns]
    pointkey_arr = [f"X{int(x*1000)}Y{int(y*1000)}" for x, y in zip(df["X_TWD97"], df["Y_TWD97"])]
    df.columns = new_cols
    df.insert(loc=0, column="PointKey", value=pointkey_arr)
    combined_output_df = pd.concat([combined_output_df, df], ignore_index=True)
```

```python
# save combined dataframe to feather for later use

combined_output_df.to_feather(os.path.join(prediction_output_folder, f"{layer}.feather"))
```

```python
unique_timepoints = combined_output_df["pred_time"].unique()
unique_timepoints[:5]
```

```python
# select_timepoint = unique_timepoints[0]
for select_timepoint in tqdm(unique_timepoints):
    df_byTime = combined_output_df.query("pred_time==@select_timepoint")
    
    df_byTime = convert_to_geodata(df=df_byTime, xcoord_col="X_TWD97", ycoord_col="Y_TWD97", crs_epsg="EPSG:3826")

    savefolder = os.path.join(prediction_output_folder, f"{layer}_shapefiles")

    if not os.path.exists(savefolder):
        os.makedirs(savefolder, exist_ok=True)

    df_byTime.to_file(os.path.join(savefolder, f"{layer}_predict_{str(select_timepoint).zfill(3)}.shp"))
```
