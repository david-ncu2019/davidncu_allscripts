```python
fpath = r"E:\SUBSIDENCE_PROJECT_DATA\LevelingSurvey\附錄6-水準測量檢測成果\Changhua_Leveling_2024_modified.xlsx"
base = os.path.basename(fpath)

df = pd.read_excel(fpath)

df = df.set_index("ALTCODE")

info_cols = [col for col in df.columns if not col.startswith("N")]
datetime_cols = [col for col in df.columns if col.startswith("N")]

show(df)
```

```python
measure_df = df.loc[:, datetime_cols]
# measure_df.columns = pd.to_datetime([col[1:] for col in measure_df.columns])

temp = measure_df.loc[:, "N2016":]

# Drop rows where first or last column is NaN
temp = temp.dropna(subset=[temp.columns[0], temp.columns[-1]])

# Then drop rows with fewer than 4 non-NaN values (considering all columns)
temp = temp.dropna(thresh=4)

info_df = df.loc[:, info_cols]
temp = pd.concat([info_df.loc[temp.index, :], temp], axis=1)

# temp.to_pickle(base.replace(".xlsx", "2016_2024.xz"))
```

```python
gdf = geospatial.convert_to_geodata(
    df=temp, xcoord_col="X_TWD97", ycoord_col="Y_TWD97", crs_epsg="EPSG:3826"
)
gdf.to_file(base.replace(".xlsx", ".shp"))
```
