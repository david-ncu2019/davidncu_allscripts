This script is used to calculate the differences between consecutive GPS measurements based on the SAR image acquisitions

```python
from appgeopy import *
from my_packages import *
```

```python
excel_fpath = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_Combined\4__Intersected\InSAR_ModelGPS_Intersect_v2.xlsx"
available_sheetnames = data_io.get_sheetnames(excel_fpath)
available_sheetnames[:5]
```

```python
savefolder = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_Combined\5__Differencing"
savename = "InSAR_ModelGPS_Differencing_v1.xlsx"
savepath = os.path.join(savefolder, savename)
savepath
```

```python
# select_station = available_sheetnames[0]

for select_station in tqdm(available_sheetnames):
    df = pd.read_excel(
        excel_fpath, sheet_name=select_station, parse_dates=[0], index_col=[0]
    )
    
    difference_in_mm = (df - df.shift(1)) * 10
    difference_in_mm = difference_in_mm.drop(labels=df.index[0])
    difference_in_mm.columns = [
        name.replace("(cm)", "(mm)") for name in difference_in_mm.columns
        # name.replace("(m)", "(mm)") for name in difference_in_mm.columns
    ]
    difference_in_mm = difference_in_mm.reset_index(drop=False)
    
    data_io.save_df_to_excel(
        df_to_save=difference_in_mm,
        filepath=savepath,
        sheet_name=select_station,
        index=False,
        verbose=False,
    )
```
