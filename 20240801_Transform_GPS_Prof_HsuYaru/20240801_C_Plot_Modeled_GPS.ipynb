```python
topfolder = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_Combined\3__Modeled"
target_filename = r"GPS_Merged_Modeled_v3.xlsx"

folder2savefig = os.path.join(topfolder, "Figures_v3")
os.makedirs(folder2savefig, exist_ok=True)

fpath = os.path.join(topfolder, target_filename)
fpath
```

```python
available_sheetnames = data_io.get_sheetnames(fpath)
available_sheetnames[:5]
```

```python
for select_sheet in tqdm(available_sheetnames):

    df_bySheetName = pd.read_excel(fpath, parse_dates=[0], index_col=[0], sheet_name=select_sheet)
    station_name = select_sheet
    
    fig, axes = visualize.plot_three_subplots(
        df_bySheetName,
        linestyle=" ",
        marker="o",
        alpha=0.2,
        suptitle=station_name,
        colors=["red", "limegreen", "deepskyblue"],
        figsize=(11.7, 8.3 * 2 / 3),
    )
    
    for ax, col in zip(axes, df_bySheetName.columns[3:]):
        ax.plot(df_bySheetName[col], color='black', lw=1.5, linestyle='-', label="Modeled(cm)")
        ax.legend(frameon=False)
        months = mdates.MonthLocator(interval=1)
        monthyearFmt = mdates.DateFormatter("%Y/%m")
        ax.xaxis.set_major_locator(months)
        ax.grid(which="major", color="lightgrey", axis="x")
        # start_time = datetime(processing_year, 1, 1) - timedelta(days=15)
        # end_time = datetime(processing_year, 12, 31) + timedelta(days=15)
        start_time = datetime(2022, 1, 1) - timedelta(days=15)
        end_time = datetime(2023, 12, 31) + timedelta(days=15)
        ax.set_xlim(start_time, end_time)

    fig_savepath = os.path.join(folder2savefig, f"{station_name}_model.png")
    visualize.save_figure(fig, fig_savepath, dpi=300)
    plt.close()
```
