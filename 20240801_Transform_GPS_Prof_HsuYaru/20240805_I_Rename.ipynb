```python
fpath = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_Combined\6__Merged\Merged_Differencing_LOS_v2.xlsx"
available_sheetnames = data_io.get_sheetnames(fpath)
available_sheetnames[:5]
```

```python
select_sheet = available_sheetnames[0]

savefolder = os.path.dirname(fpath)
savename = "Merged_Differencing_LOS_v2_ColModified.xlsx"
savepath = os.path.join(savefolder, savename)

for select_sheet in tqdm(available_sheetnames):

    df = pd.read_excel(fpath, sheet_name=select_sheet)
    old_columns = df.columns
    new_columns = [col.split("_")[-1] for col in old_columns]
    
    df = df.rename({old_col:new_col for old_col, new_col in zip(old_columns, new_columns)}, axis=1)

    data_io.save_df_to_excel(df_to_save=df, filepath=savepath, sheet_name=select_sheet, verbose=False)
```
