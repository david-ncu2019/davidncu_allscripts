```python
FILE_FOLDER = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_HsuYaru\2__Processed\A__Fulltime"
SAVE_FOLDER = r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_HsuYaru\2__Processed\B__GPS_Oct2022__2024"

FILES = glob(FILE_FOLDER+"\\*.csv")
FILES[:5]
```

```python
# fpath = FILES[0]

cache = {
    "STATION":[],
    "START":[],
    "END":[],
    "LENGTH":[]
}
for fpath in tqdm(FILES):
    df = pd.read_csv(fpath, parse_dates=[0], index_col=[0])
    extract_df = df.loc["2022":, :]
    if len(extract_df)>0:
        basename = os.path.basename(fpath)
        savepath = os.path.join(SAVE_FOLDER, basename.replace(".csv", ".xlsx"))
        start_date = extract_df.first_valid_index()
        end_date = extract_df.last_valid_index()
        cache["STATION"].append(basename.split(".")[0])
        cache["START"].append(start_date.strftime('%Y-%m-%d'))
        cache["END"].append(end_date.strftime('%Y-%m-%d'))
        cache["LENGTH"].append((end_date-start_date).days)
        extract_df = extract_df.reset_index(drop=False)
        extract_df.to_excel(savepath, index=False)

```

```python
pd.DataFrame(cache).sort_values(by="LENGTH", ascending=True).to_excel(r"E:\SUBSIDENCE_PROJECT_DATA\GPS_2024_HsuYaru\2__Processed\Extracted_Data_Summary.xlsx", index=False)
```
