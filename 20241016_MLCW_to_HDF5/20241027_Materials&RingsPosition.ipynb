```python
import pinyin
from appgeopy import *
from my_packages import *
import plot_bh

pd.set_option("display.max_columns", 500)
pd.set_option("display.max_rows", 500)
```

```python
mlcw_hdf5_file = r"20241023_MLCW_CRFP_v4.h5"

# Extract existing data and metadata
with h5py.File(mlcw_hdf5_file, "r") as hdf5_file:
    existing_data_dict = gwatertools.h5pytools.hdf5_to_data_dict(hdf5_file)
    existing_metadata_dict = gwatertools.h5pytools.hdf5_to_metadata_dict(hdf5_file)
```

```python
folder = r"D:\1000_SCRIPTS\001_PreQE_Scripts\MultiLayerCompactionMonitoringWells\MLCW_Data_Extraction"
files2process = glob(os.path.join(folder, "*.xlsx"))
files2process[:5]
```

```python
for mat_fpath in tqdm(files2process[:]):
    try:
        basename = os.path.basename(mat_fpath)
        ename = basename.split("_")[2]
        material_table = pd.read_excel(mat_fpath)
        
        depth_arr = existing_data_dict[ename]['depth']
        
        fig = plt.figure(figsize=(8.3, 11.7))
        
        ax1 = fig.add_subplot(121)
        plot_bh.plot_original_material(ax1, material_table)
        # ax1.plot(np.zeros(len(depth_arr)), depth_arr*-1, marker='_', markersize=15, linestyle=' ', color='black')
        for d in depth_arr:
            ax1.plot(0, -d, marker='x', markersize=5, linestyle=' ', color='black')
            ax1.axhline(y=-d, xmax=0.5, color='blue', linestyle='--', lw=0.75)
        
        ax2 = fig.add_subplot(122)
        plot_bh.plot_classified_material(ax2, material_table)
        # ax2.plot(np.zeros(len(depth_arr)), depth_arr*-1, marker='_', markersize=15, linestyle=' ', color='black')
        for d in depth_arr:
            ax2.plot(0, -d, marker='x', markersize=5, linestyle=' ', color='black')
            ax2.axhline(y=-d, xmax=0.5, color='blue', linestyle='--', lw=0.75)
        
        fig.suptitle(t=ename, fontsize=24, fontweight="bold", y=0.925)
        
        plt.subplots_adjust(wspace=0.1, hspace=0.1)
        visualize.save_figure(fig, savepath=os.path.join(f"Figures\{ename}.png"), dpi=1200)
        plt.close()
    except Exception as e:
        print(ename, e)
        pass
```
