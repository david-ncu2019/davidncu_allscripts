```python
def convert_MATLABtime_2string(matlab_datenum):
    """
    Convert a MATLAB datenum to a Python datetime object.
    
    Parameters:
    - matlab_datenum (float): MATLAB datenum value (e.g., 739085)
    
    Returns:
    - datetime.datetime: Corresponding Python datetime object
    """
    # MATLAB's epoch reference date (0000-01-01)
    matlab_epoch = datetime(1, 1, 1)
    
    # Subtract 366 days to align MATLAB datenum with Python's Gregorian calendar
    adjusted_days = matlab_datenum - 366
    
    # Convert to Python datetime
    python_datetime = matlab_epoch + timedelta(days=adjusted_days)
    
    return python_datetime.strftime("%Y%m%d")
```
