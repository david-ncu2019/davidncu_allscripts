#### search PS points surrounding GPS stations

```python
# convert coordinate system of PS points to GPS
transformed_TSX_PS_gdf = TSX_PS_gdf.to_crs(GPS_gdf.crs)
transformed_TSX_PS_gdf.crs
```

```python
print("Finding nearby PSCs for each GPS station...")
radius_of_buffer = 500
cache = []

for idx, row in tqdm(
    GPS_gdf.iterrows(), total=GPS_gdf.shape[0], desc="Processing GPS stations"
):
    try:
        _temp = geospatial.find_point_neighbors(
            central_point=row,
            target_points_gdf=transformed_TSX_PS_gdf,
            central_key_column="STATION",
            buffer_radius=radius_of_buffer,
        )
        cache.append(_temp)
    except Exception as e:
        print(f"Error processing station {row['STATION']}: {e}")
```

```python
# Combine all the collected results into a single DataFrame
nearby_psc_pointkey = nearby_psc.loc[:, ["PointKey", "LON", "LAT", "STATION"]].reset_index(drop=True)

prefix = "TSX"
# Construct the output filename and save path
output_savename = f"{prefix}_PSC_nearbyGPS_{radius_of_buffer}m.pkl"
output_savepath = os.path.join(output_savename)

# Save the resulting DataFrame to a pickle file
try:
    nearby_psc_pointkey.to_pickle(output_savepath, compression="zip")
    print(f"Saved nearby PSC data to {output_savepath}")
except Exception as e:
    print(f"Error saving nearby PSC data: {e}")
```
