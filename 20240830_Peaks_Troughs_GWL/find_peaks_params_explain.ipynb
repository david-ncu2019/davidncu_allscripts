#### Understanding `prominence`

In this updated visualization:

- The **orange dashed lines** represent the **prominence** of each peak, which is the vertical distance from the peak to the higher of the left or right base.
- The **green circles** are the **left bases**, and the **purple circles** are the **right bases**, indicating the lowest points surrounding the peak.
- The prominence is visually illustrated by how far the peak stands above the surrounding signal, with respect to these base points.

#### Understanding `distance`

In this visualization:

- The **orange dashed lines** represent the **horizontal distance** between consecutive peaks.
- The numbers annotated along the lines indicate the **number of data points (indices)** between each pair of peaks.

The visualizations for **prominence** and **distance** of troughs (local minima) are as follows:

1. **First Plot (Prominence)**:
   - The **orange dashed lines** represent the prominence of each trough, which is the vertical distance between the trough and the lower of the left or right bases.
   - The **green circles** represent the left bases, and the **purple circles** represent the right bases.

2. **Second Plot (Distance)**:
   - The **orange dashed lines** represent the horizontal distance between consecutive troughs.
   - The numbers along the lines show the distance between each pair of consecutive troughs in terms of the number of data points.
