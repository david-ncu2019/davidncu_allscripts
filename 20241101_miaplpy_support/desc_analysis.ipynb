```python
def plot_histogram(array, bins=30, output_file=None):
    """
    Plots a histogram for all non-zero values in a 2D NumPy array.
    
    Parameters:
    - array: A 2D NumPy array with numerical values.
    - bins: Number of bins for the histogram (default is 30).
    - output_file: If specified, saves the plot to the given file path.
    """
    # Validate input
    if not isinstance(array, np.ndarray) or array.ndim != 2:
        raise ValueError("Input must be a 2D NumPy array.")
    
    # Exclude zero values
    non_zero_values = array[array != 0]
    
    # Plot histogram
    plt.figure(figsize=(8, 6))
    sns.histplot(non_zero_values, bins='auto', kde=False, color='blue', edgecolor='black', stat='percent')
    
    # Customize plot for scientific publication
    plt.title("Histogram of Non-Zero Values", fontsize=16)
    # plt.xlabel("Value", fontsize=14)
    # plt.ylabel("Frequency", fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    
    # # Save or show plot
    # if output_file:
    #     plt.savefig(output_file, dpi=300)
    # else:
    #     plt.show()
```
