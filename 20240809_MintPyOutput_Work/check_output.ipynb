```python
fpath = r"D:\003_PRESENTATIONS\20_TSX_Sentinel_Report\scatterplot\S1A_TSX_AvgVelo_Compare.pkl"
df = pd.read_pickle(fpath, compression="zip")
df = df.dropna(how="any").reset_index(drop=True)
df.head(5)
```

```python
ratio = abs(df["velocity_mm_per_year"] / df["averaged_measurement"])
df["S1A_TSX_Ratio"] = ratio
```

```python
columns = df.columns.to_list()
new_columns = columns[:-3] + [
    "averaged_measurement",
    "S1A_TSX_Ratio",
    "geometry",
]
new_df = df.loc[:, new_columns]
new_df.head(5)
```

```python
new_df = new_df.loc[
    :,
    [
        "POINT_X",
        "POINT_Y",
        "PointKey",
        "velocity_mm_per_day",
        "velocity_mm_per_year",
        "averaged_measurement",
        "S1A_TSX_Ratio",
        "geometry"
    ],
]
```

```python
select_array = new_df["S1A_TSX_Ratio"]
q1 = np.quantile(select_array, 0.25)
q3 = np.quantile(select_array, 0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr

filtered_df = new_df[(new_df["S1A_TSX_Ratio"]>=lower_bound)&(new_df["S1A_TSX_Ratio"]<=upper_bound)]
```
