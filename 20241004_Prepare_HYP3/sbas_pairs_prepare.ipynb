```python
df_1 = pd.read_excel(r"D:\Downloads\descending_sbas_2014_2024_v1.xlsx")
df_2 = pd.read_excel(r"D:\Downloads\descending_sbas_2014_2024_v2.xlsx")
```

```python
df_1.columns = [col.strip() for col in df_1.columns]
df_2.columns = [col.strip() for col in df_2.columns]
```

```python
df_1.head(5)
```

```python
def extract_date(sceneName):
    sceneName_split = sceneName.split("_")
    sceneName_split_datestring = sceneName_split[5]
    extract_sceneName_date = sceneName_split_datestring.split("T")[0]
    return extract_sceneName_date

def get_pairs(dataframe):
    reference_array = dataframe['Reference']
    secondary_array = dataframe['Secondary']
    
    ref_date = reference_array.apply(extract_date)
    sec_date = secondary_array.apply(extract_date)
    
    dataframe['Ref_Date'] = pd.to_datetime(ref_date)
    dataframe['Second_Date'] = pd.to_datetime(sec_date)
    
    pairs = [f"r{ref}_d{sec}" for ref, sec in zip(ref_date, sec_date)]
    dataframe['Pairs'] = pairs
    return dataframe
```

```python
df_1 = get_pairs(df_1)
df_2 = get_pairs(df_2)
```

```python
df_combined = pd.concat([df_1, df_2], ignore_index=True)
df_combined = df_combined.sort_values(by='Pairs')
df_combined = df_combined.drop_duplicates(subset='Pairs', keep='first').reset_index(drop=True)
df_combined = df_combined.drop(['Reference URL', 'Secondary URL'], axis=1)
```

```python
df_combined.to_excel(r"D:\SharedFolder\support_files\SBAS_descending_pairs.xlsx", index=False)
```
