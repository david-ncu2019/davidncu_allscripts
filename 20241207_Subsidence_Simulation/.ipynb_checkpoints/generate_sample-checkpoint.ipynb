```python
from appgeopy import *
from my_packages import *
```

```python
gwl_array = synthetic_daily_signal(
    start_date="2019-01-01",
    end_date="2024-12-31",
    linear_slope=0,
    amplitude_list=[0.5, 1],
    period_list=[1, 0.5],
    variance=0.05,
)
deformation_arr = synthetic_daily_signal(
    start_date="2019-1-01",
    end_date="2024-12-31",
    linear_slope=0,
    amplitude_list=[0.4, 0.7],
    period_list=[1., 0.5],
    variance=0.075,
)
deformation_arr = deformation_arr.shift(12)
deformation_arr = deformation_arr.dropna()
fig = plt.figure(figsize=(12,3))
ax = fig.add_subplot(111)
gwl_array.plot(ax=ax, color='black')
deformation_arr.plot(ax=ax, color='blue')
```

```python
mutual_index = gwl_array.index.intersection(deformation_arr.index)
df = pd.DataFrame(data=None, index=mutual_index)
df['GWL'] = df.index.map(gwl_array['value'])
df['SUB'] = df.index.map(deformation_arr['value'])
```

```python
df.to_csv("GWL_SUB_table.csv")
```
