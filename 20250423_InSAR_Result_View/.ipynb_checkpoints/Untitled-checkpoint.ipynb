read the monthly vertical cumdisp file

```python
yunlin_points = pd.read_pickle(r"yunlin_sbasinsar/monthly_vert_yunlin.pkl", compression="zip")
yunlin_points.head(5)
```

```python
yunlin_velocities = pd.read_pickle(
    r"yunlin_sbasinsar/Yearly_AverageVelocity_Yunlin.pkl", compression="zip"
)
yunlin_velocities["X_TWD97"] = yunlin_velocities.index.map(yunlin_points.X_TWD97)
yunlin_velocities["Y_TWD97"] = yunlin_velocities.index.map(yunlin_points.Y_TWD97)
yunlin_velocities = geospatial.convert_to_geodata(
    df=yunlin_velocities, xcoord_col="X_TWD97", ycoord_col="Y_TWD97", crs_epsg="EPSG:3826"
)
yunlin_velocities.to_pickle(
    r"yunlin_sbasinsar/Yearly_AverageVelocity_Yunlin.pkl", compression="zip"
)
```

```python
huwei_points = pd.read_pickle(r"huwei_sbasinsar/monthly_vert_huwei.pkl", compression="zip")
huwei_points.head(5)
```

```python
huwei_velocities = yunlin_velocities.loc[huwei_points.index, :]
huwei_velocities.to_pickle("huwei_sbasinsar/Yearly_AverageVelocity_Huwei.pkl", compression="zip")
```

```python
sciencepark_polygon = gpd.read_file("shapefiles/huwei_science_park.shp")

temp = geospatial.find_points_within_polygon(
    points_gdf=huwei_velocities, polygon_geom=sciencepark_polygon.geometry
)

temp.to_pickle(
    "huwei_sciencepark_sbasinsar/Yearly_AverageVelocity_SciencePark.pkl", compression="zip"
)
```
