```python
original_fpath = "yunlin_sbasinsar/monthly_vert_yunlin.pkl"
original_df = pd.read_pickle(original_fpath, compression='zip')

trend_fpath = "Polynomial3_TrendSeries_Yunlin.pkl"
trend_df = pd.read_pickle(trend_fpath, compression='zip')
```

```python
original_df.head(3), trend_df.head(3)
```

```python
info_df = original_df.iloc[:, :2]
original_measure_df = original_df.iloc[:, 2:-1]
original_measure_df.head(5)
```

```python
detrend_df = original_measure_df-trend_df
detrend_df.head(5)
```

```python
detrend_df_addInfo = pd.concat([info_df, detrend_df, original_df.loc[:, ["geometry"]]], axis=1)
detrend_df_addInfo.to_pickle(r"Fulltime_Detrend_Poly3_Yunlin.pkl", compression='zip')
```

```python
from sklearn.preprocessing import MinMaxScaler

df_T = detrend_df.T

scaler = MinMaxScaler(feature_range=(-1, 1))

scaled_data = scaler.fit_transform(df_T).T

# Create new DataFrame with same index/columns
scaled_df = pd.DataFrame(
    scaled_data, 
    index=detrend_df.index, 
    columns=detrend_df.columns
)
```

```python
scaled_df_addInfo = pd.concat([info_df, scaled_df, original_df.loc[:, ["geometry"]]], axis=1)
scaled_df_addInfo = gpd.GeoDataFrame(data=scaled_df_addInfo, geometry=scaled_df_addInfo["geometry"], crs="EPSG:3826")
scaled_df_addInfo.to_pickle(r"Scaled_Detrend_Poly3_Yunlin.pkl", compression='zip')
```
