#### cumdisp to displacement

```python
cumdisp_fpath = r"D:\1000_SCRIPTS\003_Project002\20251111_GTWR003\1_PrepareDatasets\Leveling\CORRECTED_Monthly_CUMDISP_saveqgis_Oct2025_updateNov.xz"
cumdisp_df = pd.read_pickle(cumdisp_fpath)
cumdisp_df = cumdisp_df.reset_index(drop=False)
```

```python
measure_cols = [col for col in cumdisp_df.columns if col.startswith("D")]
measure_df = cumdisp_df.loc[:, measure_cols]
measure_difference = measure_df.diff(axis=1)
measure_difference.iloc[:, 0] = measure_df.iloc[:, 0]
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
new_cols = []
for col in tqdm(measure_difference.columns):
    str2date = datetime.strptime(col[1:], "%Y%m%d")
    new_date = str2date + relativedelta(days=1)
    new_date2str = "N" + new_date.strftime("%Y%m%d")
    new_cols.append(new_date2str)

measure_difference.columns = new_cols
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
final_disp_output = pd.concat(
    [cumdisp_df.iloc[:, :3], measure_difference], axis=1
)

show(final_disp_output)
```

```python
final_disp_output.to_feather("CORRECTED_Monthly_DISPLACEMENT_updateNov.feather")
```
