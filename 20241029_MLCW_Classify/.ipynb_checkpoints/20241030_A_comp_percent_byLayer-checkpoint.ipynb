```python
def getMLCW_byStation(data_dict, station_name, value_type):
    depth_arr = data_dict[station_name]["depth"]
    date_arr = [datetime_handle.bytes_to_datetime(ele) for ele in data_dict[station_name]["date"]]
    compaction_arr = data_dict[station_name]["values"][value_type]
    df = pd.DataFrame(data=compaction_arr, index=depth_arr, columns=date_arr)
    return df
```

```python
def differ_to_ref(series, convert_to=None):
    """
    Calculate differential values relative to the first measurement in the series.

    Parameters:
    - series (pd.Series): A pandas Series containing the measurements.
    - convert_to (str or None): Unit to convert the differential values to.
                                Accepts 'milimeter', 'centimeter', or None for 'meters'.

    Returns:
    - np.ndarray: The differential values converted to the specified unit.

    Raises:
    - ValueError: If 'convert_to' is not 'milimeter', 'centimeter', or None.

    """
    # Determine the conversion multiplier based on the desired unit.
    if convert_to == "milimeter":
        multiplier = 1000
    elif convert_to == "centimeter":
        multiplier = 100
    elif convert_to is None:
        multiplier = 1  # Keep the original unit (meters).
    else:
        raise ValueError("Invalid 'convert_to' value. Must be 'milimeter', 'centimeter', or None for meters.")

    # Calculate differential values relative to the first measurement and apply conversion.
    return np.array((series - series.iloc[0]) * multiplier, dtype=np.float16)
```

```python
# -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- -- --
def cdisp_from_base(input_array):
    cdisp_ref_to_base = np.nancumsum(input_array[::-1], dtype=np.float64)
    return cdisp_ref_to_base[::-1]
```

```python
def get_jet_colors(n, vmin=0, vmax=1, cmap_name="jet"):
    cmap = plt.get_cmap(cmap_name)
    norm = Normalize(vmin=vmin, vmax=vmax)  # Normalize the color range from vmin to vmax
    colors = [cmap(norm(i)) for i in np.linspace(vmin, vmax, n)]
    return colors
```

```python
# Path to the folder containing magnetic ring layers
magneticring_fld = r"E:\SUBSIDENCE_PROJECT_DATA\2024_MLCW_ANALYSIS\megnaticRingsLayer"

# Load magnetic ring metadata from a table containing Chinese station names and well codes
well_info_table = pd.read_table(
    os.path.join(magneticring_fld, "1234.txt"),
    header=None,
    sep="\s+",
    usecols=[0, 1],
    names=["CNAME", "WELLCODE"],
)
well_info_table.head(5)  # Display the first few rows for verification
```

```python
# --------------------------------------------------------------------------------
# Calculate Cumulative Compaction by Station
# --------------------------------------------------------------------------------
```

```python
# List of all stations available in the data dictionaries (from HDF5)
stations = [
    word for word in sorted(set(existing_data_dict.keys()).union(set(existing_metadata_dict.keys())))
    if word.isupper()
]

# Select a target station to process
select_station = "TUKU"

# Extract the cumulative compaction data for the selected station
column_diffs = getMLCW_byStation(data_dict=existing_data_dict, station_name=select_station, value_type="col_diff")
column_diffs = (column_diffs * 100).transpose()  # Convert to centimeters and transpose

# Forward-fill and backward-fill NaN values for continuous time series
filled_diffs = column_diffs.fillna(method="ffill").fillna(method="bfill")

# Compute differential values relative to the first measurement
diffs_ref2first_transposed = filled_diffs.apply(lambda x: differ_to_ref(x), axis=0).transpose()

# Calculate cumulative compaction by summing values from the base of the borehole upwards
cdisp_ref2base = diffs_ref2first_transposed.apply(cdisp_from_base, axis="index")
# Extract the cumulative compaction array (the top array)
ref2base_cumpcomp_arr = cdisp_ref2base.iloc[0, :]
# # ---------------------------------------------------------------------------------------------------

# get the Chinese name from station's metadata
cname_byStation = existing_metadata_dict[select_station]["LandSubsidenceMonitoringWellName"]

# from the chinese name, we get the wellcode, for example LNJS, TKJS,...
wellcode_byStation = well_info_table.query("CNAME==@cname_byStation")["WELLCODE"].iloc[0]
```

```python
# Retrieve the station's Chinese name from the metadata dictionary
cname_byStation = existing_metadata_dict[select_station]["LandSubsidenceMonitoringWellName"]

# Find the corresponding well code from the well info table using the Chinese name
wellcode_byStation = well_info_table.query("CNAME == @cname_byStation")["WELLCODE"].iloc[0]
```

```python
# --------------------------------------------------------------------------------
# Load Magnetic Ring Layer Data for the Well Code
# --------------------------------------------------------------------------------
```

```python
# Path to the Excel file containing layer information for the selected well code
magneticring_fpath = os.path.join(magneticring_fld, wellcode_byStation + ".xlsx")

# Load the layer information from the Excel file
magneticring_df = pd.read_excel(magneticring_fpath)

# Extract available unique layer names within the well data
available_layers = magneticring_df["Layer"].unique()
```

```python
# --------------------------------------------------------------------------------
# Calculate Cumulative Compaction by Each Layer
# --------------------------------------------------------------------------------
```

```python
cum_comp_allLayer = pd.DataFrame()

# Loop through each layer and calculate cumulative compaction for that specific depth range
for select_layer in available_layers:
    ring_byLayer = magneticring_df.query("Layer == @select_layer").iloc[:, 1]
    min_depth, max_depth = ring_byLayer.min(), ring_byLayer.max()

    # Filter data within the depth range of the selected layer and compute cumulative compaction
    ref2first_byLayer = diffs_ref2first_transposed.query("@min_depth <= index <= @max_depth")
    cum_comp_byLayer = ref2first_byLayer.apply(cdisp_from_base, axis="index")

    # Select the first row for the cumulative compaction of each layer and update the index
    cum_comp_byLayer_arr = cum_comp_byLayer.iloc[[0], :]
    cum_comp_byLayer_arr.index = [select_layer]

    # Concatenate the results for each layer into a single DataFrame
    cum_comp_allLayer = pd.concat([cum_comp_allLayer, cum_comp_byLayer_arr], ignore_index=False)

# Transpose to make layers as columns for easier plotting
cum_comp_allLayer = cum_comp_allLayer.T
```

```python
# --------------------------------------------------------------------------------
# Fill Missing Time- Points and Interpolate Missing Values
# -------------------------------------------------------------------------------
```

```python
# Generate a complete time series covering the full range of recorded dates
fulltime_arr = datetime_handle.get_fulltime(cum_comp_allLayer.index)

# Create a DataFrame with a full-time index and interpolate missing cumulative compaction values
fulltime_cum_comp = datetime_handle.fulltime_table(df=cum_comp_allLayer, fulltime_series=fulltime_arr)
filled_fulltime_cum_comp = fulltime_cum_comp.interpolate(method="from_derivatives")
```
