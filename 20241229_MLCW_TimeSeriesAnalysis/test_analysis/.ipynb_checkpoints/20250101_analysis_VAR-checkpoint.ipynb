```python
"""
Program: Enhanced VAR Analysis and Prediction with Error Handling
Description: Robust analysis and prediction of worker productivity using Vector Autoregression (VAR).
Includes handling for missing values, optimal lag selection, and improved evaluation metrics.
Requirements: Install `statsmodels`, `numpy`, `pandas`, `matplotlib` via pip.
"""

import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller

def load_data(file_path):
    """
    Load the synthetic worker productivity dataset from a CSV file.

    Parameters:
        file_path (str): Path to the CSV file containing the dataset.

    Returns:
        pd.DataFrame: A DataFrame containing the productivity data.
    """
    data = pd.read_csv(file_path, index_col=0, parse_dates=True)
    data = data.dropna()  # Drop missing values
    data.index.freq = pd.infer_freq(data.index)  # Ensure frequency information is set
    return data

def check_stationarity(data):
    """
    Perform the Augmented Dickey-Fuller test to check stationarity.

    Parameters:
        data (pd.DataFrame): Dataset to test.

    Returns:
        bool: True if stationary, False otherwise.
    """
    p_values = []
    for col in data.columns:
        result = adfuller(data[col])
        p_values.append(result[1])
    return all(p < 0.05 for p in p_values)

def make_stationary(data):
    """
    Convert data to stationary by differencing if needed.

    Parameters:
        data (pd.DataFrame): Input time series data.

    Returns:
        pd.DataFrame: Differenced (stationary) data.
    """
    return data.diff().dropna()

def split_data(data, train_end_date):
    """
    Split the dataset into training and testing sets based on a cutoff date.

    Parameters:
        data (pd.DataFrame): The complete dataset.
        train_end_date (str): The end date for the training set in 'YYYY-MM-DD' format.

    Returns:
        pd.DataFrame, pd.DataFrame: Training and testing datasets.
    """
    train_data = data.loc[:train_end_date]
    test_data = data.loc[train_end_date:]
    return train_data, test_data

def fit_var_model(train_data, max_lags=15):
    """
    Fit a VAR model to the training dataset.

    Parameters:
        train_data (pd.DataFrame): The training dataset.
        max_lags (int): Maximum lag order for the VAR model.

    Returns:
        VARResults: Fitted VAR model.
    """
    model = VAR(train_data)
    selected_lag = model.select_order(maxlags=max_lags)
    best_lag = selected_lag.aic  # Use AIC to select optimal lag
    print(f"Optimal Lag Order: {best_lag}")
    results = model.fit(best_lag)
    return results

def forecast_and_evaluate(model, train_data, test_data):
    """
    Forecast and evaluate the VAR model on the test dataset.

    Parameters:
        model (VARResults): The fitted VAR model.
        train_data (pd.DataFrame): The training dataset.
        test_data (pd.DataFrame): The testing dataset.

    Returns:
        pd.DataFrame, dict: Predicted values and evaluation metrics (RMSE, MAE, MAPE).
    """
    forecast = model.forecast(train_data.values[-model.k_ar:], steps=len(test_data))
    forecast_df = pd.DataFrame(forecast, index=test_data.index, columns=test_data.columns)

    # Ensure no NaN values
    forecast_df = forecast_df.fillna(0)

    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(test_data, forecast_df))
    mae = mean_absolute_error(test_data, forecast_df)
    mape = np.mean(np.abs((test_data - forecast_df) / test_data)) * 100

    metrics = {"RMSE": rmse, "MAE": mae, "MAPE": mape}
    return forecast_df, metrics

def plot_results(test_data, predictions, metrics):
    """
    Plot actual vs predicted values and residuals.

    Parameters:
        test_data (pd.DataFrame): Actual test dataset.
        predictions (pd.DataFrame): Predicted values.
        metrics (dict): Evaluation metrics.
    """
    plt.figure(figsize=(12, 8))

    # Plot test data (actual values)
    for col in test_data.columns:
        plt.plot(test_data.index, test_data[col], label=f"Actual {col}", linestyle="--", linewidth=2)

    # Plot predictions
    for col in predictions.columns:
        plt.plot(predictions.index, predictions[col], label=f"Predicted {col}", linewidth=2)

    # Add labels, title, and grid
    plt.title(f"VAR Forecast vs. Actual\nRMSE: {metrics['RMSE']:.2f}, MAE: {metrics['MAE']:.2f}, MAPE: {metrics['MAPE']:.2f}%", fontsize=16, fontweight='bold')
    plt.xlabel("Date", fontsize=14)
    plt.ylabel("Productivity", fontsize=14)
    plt.grid(visible=True, linestyle="--", linewidth=0.5, alpha=0.7)

    # Format x-axis for dates
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    plt.gcf().autofmt_xdate()  # Rotate date labels

    # Enhance tick parameters
    plt.tick_params(axis="both", which="major", labelsize=12)

    # Add legend outside the plot
    plt.legend(fontsize=12, loc="upper left", bbox_to_anchor=(1.05, 1), borderaxespad=0.)

    # Adjust layout and show
    plt.tight_layout()
    plt.show()

    # Residual plot
    residuals = test_data - predictions
    residuals.plot(figsize=(12, 6), title="Residuals", linewidth=1.5)
    plt.axhline(0, color="red", linestyle="--", linewidth=0.8)
    plt.xlabel("Date", fontsize=14)
    plt.ylabel("Residuals", fontsize=14)
    plt.grid(visible=True, linestyle="--", linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Load the synthetic dataset
    data = load_data("synthetic_worker_productivity.csv")

    # Check stationarity and make stationary if needed
    if not check_stationarity(data):
        print("Data is not stationary. Differencing...")
        data = make_stationary(data)

    # Split into training (first 2 years) and testing (last year)
    train_data, test_data = split_data(data, "2022-12-31")

    # Fit VAR model
    var_model = fit_var_model(train_data)

    # Forecast and evaluate
    predictions, metrics = forecast_and_evaluate(var_model, train_data, test_data)

    # Plot results
    plot_results(test_data, predictions, metrics)
```
